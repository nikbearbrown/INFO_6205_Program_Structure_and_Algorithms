Find the Shortest Path in a Weighted Directed Graph
Problem Statement:
You are given a weighted directed graph and two distinct vertices, a source vertex S, and a destination vertex D. Your task is to find the shortest path from S to D in the given graph.

Input Format:

An integer N representing the number of vertices in the graph (1 <= N <= 10^5).
An integer M representing the number of directed edges in the graph (1 <= M <= 10^5).
N lines containing two integers each, U and V (1 <= U, V <= N), representing a directed edge from vertex U to vertex V.
M lines containing three integers each, U, V, and W (1 <= U, V <= N, 1 <= W <= 1000), representing the weight of the directed edge from vertex U to vertex V.
The source vertex S and destination vertex D are provided separately as two integers, where 1 <= S, D <= N and S ≠ D.

Output Format:

An integer representing the length of the shortest path from S to D. If there is no path from S to D, output -1.

Constraints:

The weight of each edge is between 1 and 1000.
The source and destination vertices are distinct.
The graph may not be strongly connected, i.e., there may not always be a path from S to D.

Solution:

To solve this problem, we can use Dijkstra's algorithm to find the shortest path from the source vertex S to all other vertices in the graph. Then, we can return the distance from S to the destination vertex D.

Here's a pseudocode solution:

function shortest_path(graph, S, D):
    Initialize a distance array dist with infinity for all vertices except S (dist[S] = 0).
    Create a priority queue Q to store vertices based on their distance.
    Enqueue S into Q with a distance of 0.
    
    while Q is not empty:
        vertex = dequeue a vertex with the minimum distance from Q
        for each neighbor of vertex:
            new_distance = dist[vertex] + weight of edge from vertex to neighbor
            if new_distance < dist[neighbor]:
                dist[neighbor] = new_distance
                enqueue neighbor into Q with new_distance
    
    if dist[D] is still infinity, return -1 (no path exists from S to D)
    else return dist[D]

The correctness of Dijkstra's algorithm is well-established and proven. It guarantees finding the shortest path from the source vertex to all other vertices in the graph.

Solution:
To encode the song efficiently, we can use the following pseudocode:

def encode_folk_song(song_lines):

    encoded_script = []
    unique_lines = set()
    
    for line in song_lines:
        if line not in unique_lines:
            encoded_script.append(line)
            unique_lines.add(line)
    
    return encoded_script


This solution iterates through the song lines and adds each line to the encoded_script only if it has not been encountered before. Since folk songs reuse lines from previous verses, we eliminate duplicates, resulting in the shortest possible script. This approach is efficient and correct because it utilizes the property of these songs effectively.

Solution:

function isSubsequence(S, S', n, m):
    i = 0  -> Pointer for sequence S
    j = 0  -> Pointer for sequence S'

    while i < n and j < m:
        if S[i] == S'[j]:
            j += 1  -> Move to the next element in S'
        i += 1  -> Always move to the next element in S

    - If we reached the end of S', S' is a subsequence of S
    return j == m

If it successfully reaches the end of S', it means all elements in S' have been found in S in the same relative order, satisfying the definition of a subsequence. Thus, the algorithm is correct.

Use Prim's algorithm to find a minimum spanning tree in a graph of your choice.

Constraints:

1 ≤ n ≤ 1000 (number of nodes)
0 ≤ m ≤ n*(n-1)/2 (maximum number of edges)
Solution:

Prim's algorithm is used to find a minimum spanning tree (MST) in a weighted, connected, and undirected graph. It works by greedily selecting edges that connect the MST to a new node while ensuring that the MST remains connected. Here's a step-by-step explanation of Prim's algorithm:

1. Initialize an empty set MST to represent the minimum spanning tree and a set visited to keep track of visited nodes.
2. Choose a starting node s arbitrarily and add it to the visited set.
3. Initialize a priority queue (min-heap) pq with all edges incident to s.
4. Repeat the following steps until MST contains n-1 edges (where n is the number of nodes in the graph):

   a. Pop the edge (u, v) with the smallest weight from the priority queue pq. If v is not in the visited set, add (u, v) to MST and mark v as visited by adding it to the visited set.

   b. Add all edges incident to v that connect to unvisited nodes to the priority queue pq.
5. Return the MST as the minimum spanning tree of the graph.



Proof of Correctness:

Prim's algorithm is a greedy algorithm, and it guarantees the correctness of the minimum spanning tree it produces. The algorithm selects edges with the smallest weight, ensuring that the tree remains connected while minimizing the total weight. By the end of the algorithm, it will have connected all nodes with the minimum possible weight, satisfying the properties of a minimum spanning tree.
