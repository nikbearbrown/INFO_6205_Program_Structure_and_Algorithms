```python
import copy
import time
start_time = time.time()
WesternTeam = {
    'Western_1': ['Eastern_8', 'Eastern_7', 'Eastern_6', 'Eastern_5', 'Eastern_4', 'Eastern_3', 'Eastern_2', 'Eastern_1'],
    'Western_2': ['Eastern_2', 'Eastern_4', 'Eastern_3', 'Eastern_8', 'Eastern_1', 'Eastern_5', 'Eastern_7', 'Eastern_6'],
    'Western_3': ['Eastern_2', 'Eastern_8', 'Eastern_1', 'Eastern_7', 'Eastern_6', 'Eastern_5', 'Eastern_4', 'Eastern_3'],
    'Western_4': ['Eastern_3', 'Eastern_6', 'Eastern_1', 'Eastern_8', 'Eastern_7', 'Eastern_2', 'Eastern_5', 'Eastern_4'],
    'Western_5': ['Eastern_1', 'Eastern_3', 'Eastern_5', 'Eastern_7', 'Eastern_2', 'Eastern_4', 'Eastern_6', 'Eastern_8'],
    'Western_6': ['Eastern_7', 'Eastern_6', 'Eastern_8', 'Eastern_4', 'Eastern_3', 'Eastern_1', 'Eastern_5', 'Eastern_2'],
    'Western_7': ['Eastern_4', 'Eastern_5', 'Eastern_2', 'Eastern_3', 'Eastern_6', 'Eastern_7', 'Eastern_8', 'Eastern_1'],
    'Western_8': ['Eastern_6', 'Eastern_8', 'Eastern_4', 'Eastern_1', 'Eastern_5', 'Eastern_2', 'Eastern_3', 'Eastern_7']
}

EasternTeam = {
    'Eastern_1': ['Western_8', 'Western_3', 'Western_1', 'Western_2', 'Western_6', 'Western_7', 'Western_5', 'Western_4'],
    'Eastern_2': ['Western_3', 'Western_2', 'Western_1', 'Western_4', 'Western_5', 'Western_6', 'Western_7', 'Western_8'],
    'Eastern_3': ['Western_1', 'Western_6', 'Western_5', 'Western_4', 'Western_3', 'Western_2', 'Western_7', 'Western_8'],
    'Eastern_4': ['Western_7', 'Western_4', 'Western_6', 'Western_2', 'Western_8', 'Western_5', 'Western_3', 'Western_1'],
    'Eastern_5': ['Western_6', 'Western_7', 'Western_4', 'Western_3', 'Western_2', 'Western_8', 'Western_1', 'Western_5'],
    'Eastern_6': ['Western_1', 'Western_5', 'Western_8', 'Western_4', 'Western_7', 'Western_3', 'Western_2', 'Western_6'],
    'Eastern_7': ['Western_4', 'Western_2', 'Western_6', 'Western_3', 'Western_1', 'Western_5', 'Western_7', 'Western_8'],
    'Eastern_8': ['Western_1', 'Western_5', 'Western_7', 'Western_2', 'Western_3', 'Western_6', 'Western_8', 'Western_4']
}

Western = sorted(WesternTeam.keys())
Eastern = sorted(EasternTeam.keys())


def check(matched):
    inversematched = dict((v,k) for k,v in matched.items())#
    for E,W in matched.items():
        Elikes = EasternTeam[E]
        Elikesbetter = Elikes[:Elikes.index(W)]
        Wlikes = WesternTeam[W]
        Wlikesbetter =Wlikes[:Wlikes.index(E)]
        for Wes in Elikesbetter:
            Western_East = inversematched[Wes] 
            Weslikes = WesternTeam[Wes] 
            if Weslikes.index(Western_East) > Weslikes.index(E):
                print("%s and %s tend to play against each other better than "
                      "their present rival: %s and %s, respectively"
                      % (E, Wes,W, Western_East))
                return False
        for Eas in Wlikesbetter:
            Eastern_Wes = matched[Eas]
            Easlikes = EasternTeam[Eas]
            if Easlikes.index(Eastern_Wes) > Easlikes.index(W):
                print("%s and %s tend to play against each other better than "
                      "their present rival: %s and %s, respectively"
                      % (W, Eas, E, Eastern_Wes))
                return False
    return True

def matchmaker():
    Westernfree = Western[:]
    matched  = {}
    WesternTeam2 = copy.deepcopy(WesternTeam)
    EasternTeam2 = copy.deepcopy(EasternTeam)
    while Westernfree:
        Wes = Westernfree.pop(0)
        Westernlist = WesternTeam2[Wes]
        Eas = Westernlist.pop(0)
        competitor = matched.get(Eas)
        if not competitor:
            # E's free
            matched[Eas] = Wes
            print("  %s and %s" % (Wes, Eas))
        else:
            Easternlist = EasternTeam2[Eas]
            if Easternlist.index(competitor) > Easternlist.index(Wes):
                matched[Eas] = Wes
                print("  %s dumped %s for %s" % (Eas, competitor, Wes))
                if WesternTeam2[competitor]:
                    
                    Westernfree.append(competitor)
            else:
                if Westernlist:
                    Westernfree.append(Wes)
    return matched


print('\nMatch:')
matched = matchmaker()

print('\nPairs:')
print('  ' + ',\n  '.join('%s is matched to %s' % couple
                          for couple in sorted(matched.items())))
print()
print('Match stability check PASSED'
      if check(matched) else 'Match stability check FAILED')
print("--- %s seconds ---" % (round(time.time() - start_time, 10)))
```

```python
import random

# Define the list of team names for both East and West teams
east_teams = ['Eastern_' + str(i) for i in range(1, 17)]
west_teams = ['Western_' + str(i) for i in range(1, 17)]

# Initialize empty preference tables for East and West teams
eastern_preferences = {}
western_preferences = {}

# Generate random preference tables for East and West teams
for east_team in east_teams:
    # Shuffle the list of west teams to create random preferences
    random_west_prefs = random.sample(west_teams, len(west_teams))
    eastern_preferences[east_team] = random_west_prefs

for west_team in west_teams:
    # Shuffle the list of east teams to create random preferences
    random_east_prefs = random.sample(east_teams, len(east_teams))
    western_preferences[west_team] = random_east_prefs

# Print the generated preference tables
print("Eastern Team Preferences:")
for team, prefs in eastern_preferences.items():
    print(f"{team}: {prefs}")

print("\nWestern Team Preferences:")
for team, prefs in western_preferences.items():
    print(f"{team}: {prefs}")
```
