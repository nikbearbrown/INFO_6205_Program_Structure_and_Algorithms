Therefore, the Directed Disjoint Circuits Problem is NP-complete, as the reduction from the NP-complete Hamiltonian Circuit Problem is polynomial, and the solution to the Directed Disjoint Circuits Problem can be verified in polynomial time.

<u>**Answer & Justification :**</u>

This problem can be reduced to a max-flow problem by constructing a bipartite graph and applying the max-flow algorithm. The steps are similar to the provided question:

<u>Construct a Bipartite Graph:</u>

Create a bipartite graph X ∪ Y.
Add a node in X for each edge in the original graph.
Add a node in Y for each node in the original graph.
Connect each node in X to the two nodes incident to the edge it represents in the original graph.
Assign a capacity greater than or equal to 1 to these edges.
Create a source node s connected to all nodes in X with each edge assigned a capacity of 1.
Create a sink node t connected from each node in Y, with each edge having a capacity equal to the out-degree upper bound for the corresponding node in the original graph.
Run Max-Flow:

Apply the max-flow algorithm (e.g., Ford-Fulkerson) on this network.


<u>Check Feasibility:</u>

If the max-flow value is less than the number of edges in the original undirected graph, report that it's not possible to orient the edges respecting the out-degree bounds.
Otherwise, orient the edges as follows:
Since nodes in X have exactly one unit of flow coming in, each node in X will pick one of its two outgoing edges to send this flow out.
The node to which this flow goes out is the head of this edge.
Since max-flow is equal to the number of edges in the original graph, all edges can be assigned a unique orientation, and this orientation respects the out-degree bounds on the graph.
Therefore, the Directed Graph Construction problem can be solved using the max-flow algorithm and the bipartite graph construction, ensuring that out-degree bounds are respected.

```python
import networkx as nx
import matplotlib.pyplot as plt

def construct_bipartite_graph(original_graph, out_degree_bounds):
    G = nx.DiGraph()

    # Create nodes in X for each edge in the original graph
    for edge in original_graph.edges():
        G.add_node(edge, bipartite=0)

    # Create nodes in Y for each node in the original graph
    for node in original_graph.nodes():
        G.add_node(node, bipartite=1)

    # Connect nodes in X to nodes in Y
    for edge in original_graph.edges():
        for node in edge:
            G.add_edge(edge, node, capacity=float('inf'))

    # Create source node connected to nodes in X
    source_node = 'source'
    G.add_node(source_node)
    for edge in original_graph.edges():
        G.add_edge(source_node, edge, capacity=1)

    # Create sink node connected from nodes in Y with capacity based on out-degree bounds
    sink_node = 'sink'
    G.add_node(sink_node)
    for node in original_graph.nodes():
        G.add_edge(node, sink_node, capacity=out_degree_bounds[node])

    return G

def draw_bipartite_graph(G):
    pos = nx.bipartite_layout(G, G.nodes(), align='horizontal')
    nx.draw(G, pos, with_labels=True, font_size=8, font_color='black', font_weight='bold', node_size=700, node_color='skyblue', arrowsize=20)
    plt.show()

# Example usage
original_graph = nx.Graph()
original_graph.add_edges_from([(1, 2), (2, 3), (3, 1), (3, 4)])
out_degree_bounds = {1: 1, 2: 1, 3: 2, 4: 1}

bipartite_graph = construct_bipartite_graph(original_graph, out_degree_bounds)
draw_bipartite_graph(bipartite_graph)
```
