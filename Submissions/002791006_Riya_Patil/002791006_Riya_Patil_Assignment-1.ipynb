### Q6. (15 marks)  
Consider you are part of a team conducting a comprehensive climate analysis for the United States. To achieve this, you have been provided with a dataset representing temperature data over time for various regions across the country. This data is structured as a 3D grid, where each cell represents the average temperature at a specific geographical location during a specific period. Your objective is to identify and track recurring climate hotspots in the United States. A hotspot is defined as a contiguous area with temperatures consistently higher than a given threshold T over a specified time range.
Write an algorithm to identify and track recurring climate hotspots in the temperature data.

#### Solution: 
**Input**
- A 3D grid G representing the temperature data for the United States. G[i][j][k] contains the average temperature at geographical location (i, j) during time period k.
- An integer threshold temperature T to define what constitutes a hotspot.
- An integer D representing the minimum duration (in time periods) a hotspot must persist to be considered significant.

**Output**  
Return a list of tuples, each containing:
- The geographical coordinates (latitude, longitude) of the hotspot.
- The time range during which the hotspot persisted.

**Pseudocode**  
    Initialize hotspots list

    For each latitude i:
        For each longitude j:
            For each time period k:
                If temperature at (i, j, k) exceeds threshold T:
                    Start a new hotspot
                    Initialize coordinates and time range
                    While temperature exceeds T and within time range:
                        Update hotspot coordinates and time range
                Else if hotspot duration meets minimum requirement:
                    Add hotspot details to hotspots list

    Return hotspots

**Constraints**
- The size of the grid and the number of time periods are within realistic bounds for climate data analysis.
- The temperature values are in Fahrenheit and fall within a realistic range.

**Complexity Analysis**  
- Time Complexity: The algorithm iterates through the entire 3D grid, resulting in a time complexity of O(nmt) where n, m, and t are the dimensions of the grid.
- Space Complexity: The space complexity is O(1) as we are using a fixed amount of space for variables and not storing additional data structures. The hotspots list grows in size based on the number of hotspots detected.

The python code is given below

```python
def find_hotspots(grid, threshold, min_duration):
    hotspots = []

    n = len(grid)
    m = len(grid[0])
    t = len(grid[0][0])

    def is_valid(i, j, k):
        return 0 <= i < n and 0 <= j < m and 0 <= k < t

    def explore(i, j, k, start_time):
        nonlocal hotspot_start, hotspot_end

        if is_valid(i, j, k) and grid[i][j][k] >= threshold:
            if hotspot_start is None:
                hotspot_start = (i, j, k, start_time)
            else:
                hotspot_end = (i, j, k, start_time)
        else:
            if hotspot_start is not None and hotspot_end is not None:
                hotspot_duration = hotspot_end[3] - hotspot_start[3] + 1
                if hotspot_duration >= min_duration:
                    hotspots.append((hotspot_start[:3], hotspot_end[:3], (hotspot_start[3], hotspot_end[3])))
                hotspot_start = hotspot_end = None

    for i in range(n):
        for j in range(m):
            for k in range(t):
                hotspot_start = hotspot_end = None
                start_time = k

                explore(i + 1, j, k, start_time)
                explore(i - 1, j, k, start_time)
                explore(i, j + 1, k, start_time)
                explore(i, j - 1, k, start_time)
                explore(i, j, k + 1, start_time)
                explore(i, j, k - 1, start_time)

    return hotspots

# Example Usage:
temperature_data = [
    [
        [25, 28, 31, 32],
        [30, 32, 32, 30],
        [35, 36, 32, 30],
        [32, 34, 33, 28]
    ],
    [
        [26, 29, 32, 33],
        [31, 33, 33, 31],
        [36, 37, 33, 31],
        [33, 35, 34, 29]
    ]
]

threshold_temperature = 32
min_duration_months = 1

hotspots = find_hotspots(temperature_data, threshold_temperature, min_duration_months)

# Print the identified hotspots
for i, hotspot in enumerate(hotspots, 1):
    print(f'Hotspot {i}: Top-left: {hotspot[0]}, Bottom-right: {hotspot[1]}, Time Range: {hotspot[2]}')
```

Consider a scenario in the medical field where there are n medical students and m hospital residency programs. Each student has specific preferences for the residency programs they want to join, and each program has preferences for the students they want to accept based on their academic performance, medical specialties, and research interests. Additionally, there is a critical aspect to this scenario. Each residency program offers different specializations, such as surgery, internal medicine, pediatrics, etc., and each student has their own preferences over these specialties, which may vary based on their career aspirations and medical interests.  
Write an algorithm to find a stable matching that maximizes the overall satisfaction, considering both the students preferences for residency programs and their preferences for medical specializations. Additionally, ensure that each student is matched with a program that aligns with their medical interests.

#### Solution: 
**Input**  
- A list of n medical students, each with a ranked preference list of residency programs and a ranked preference list of medical specializations.
- A list of m hospital residency programs, each with a ranked preference list of students, a list of medical specializations they are offering, along with the corresponding suitability levels for each specialization.
- A list of k medical specializations available in the program.  

**Output**  
A stable matching of students to residency programs, along with the chosen medical specializations for each student.

**Example**     
Input:  
Students: [M1, M2, M3]  
Residency Programs: [R1, R2, R3]  
Medical Specializations: [Surgery, Internal Medicine, Pediatrics]  

Student Preferences:  
- M1: R2 > R1 > R3, Surgery > Internal Medicine > Pediatrics
- M2: R1 > R2 > R3, Internal Medicine > Pediatrics > Surgery
- M3: R3 > R2 > R1, Pediatrics > Surgery > Internal Medicine

Residency Program Preferences:  
- R1: M2 > M1 > M3, [Internal Medicine, Pediatrics, Surgery], [8, 6, 4]
- R2: M1 > M3 > M2, [Surgery, Internal Medicine, Pediatrics], [9, 7, 5]
- R3: M3 > M1 > M2, [Pediatrics, Internal Medicine, Surgery], [8, 6, 4]  

Output:  
Matching:  
- M1 is matched with R2 (Specialization: Surgery)
- M2 is matched with R1 (Specialization: Internal Medicine)
- M3 is matched with R3 (Specialization: Pediatrics)  

Constraints  
- The number of medical students n and residency programs m is at most 100.
- The number of medical specializations k is at most 10.
- The suitability levels are integers in the range of 1 to 10.

The python code is given below

```python
def find_stable_matching(students, programs, specializations):
    matching = {}
    unmatched_students = set(range(len(students)))
    
    while unmatched_students:
        student_idx = unmatched_students.pop()
        student = students[student_idx]
        preferences = student['preferences']
        for program in preferences:
            if program in programs:
                specialization = student['specialization']
                if specialization in programs[program]['specializations']:
                    matching[student['name']] = (program, specialization)
                    unmatched_students.discard(student_idx)
                    programs[program]['students'].append(student['name'])
                    break
    
    return matching

# Example Usage:
students = [
    {'name': 'M1', 'preferences': ['R2', 'R1', 'R3'], 'specialization': 'Surgery'},
    {'name': 'M2', 'preferences': ['R1', 'R2', 'R3'], 'specialization': 'Internal Medicine'},
    {'name': 'M3', 'preferences': ['R3', 'R2', 'R1'], 'specialization': 'Pediatrics'},
]

programs = {
    'R1': {'students': [], 'specializations': ['Internal Medicine', 'Pediatrics', 'Surgery']},
    'R2': {'students': [], 'specializations': ['Surgery', 'Internal Medicine', 'Pediatrics']},
    'R3': {'students': [], 'specializations': ['Pediatrics', 'Internal Medicine', 'Surgery']},
}

specializations = ['Surgery', 'Internal Medicine', 'Pediatrics']

matching = find_stable_matching(students, programs, specializations)

# Print the matching
for student, (program, specialization) in matching.items():
    print(f'{student} is matched with {program} (Specialization: {specialization})')
```

#### Solution:

This algorithm selects regions with the highest population for each chain, ensuring that the locations collectively cover the highest population while minimizing overlap.

2. Let's consider a competitive scenario with specific population distributions and store allocation plans for Chain A and Chain B:
- Population Distribution: [25000, 18000, 20000, 15000, 10000]
- Chain A plans to open 3 stores.
- Chain B plans to open 2 stores.

In this scenario, both chains are competing for the same regions, and the population distribution is such that different regions have varying levels of demand. The competition for optimal locations will be intense as each chain tries to secure regions with the highest population and demand.

After running the algorithm or analyzing various options, it may be determined that there is no clear-cut stable solution. The choice of locations for one chain significantly affects the optimal choices for the other chain, leading to a competitive scenario with no clear stable outcome.


```python
def find_optimal_locations(population_distribution, num_stores_A, num_stores_B):
    total_population = sum(population_distribution)
    regions = len(population_distribution)

    # Initialize store locations for both chains
    locations_A = []
    locations_B = []

    for _ in range(num_stores_A):
        # Find the region with the highest population
        max_population_index = population_distribution.index(max(population_distribution))
        # Add the region to Chain A's locations
        locations_A.append(max_population_index)
        # Reduce the population of the chosen region to avoid re-selection
        population_distribution[max_population_index] = -1

    for _ in range(num_stores_B):
        # Find the region with the highest remaining population
        max_population_index = population_distribution.index(max(population_distribution))
        # Add the region to Chain B's locations
        locations_B.append(max_population_index)
        # Reduce the population of the chosen region to avoid re-selection
        population_distribution[max_population_index] = -1

    return locations_A, locations_B

# Example Usage:
population_distribution = [10000, 20000, 15000, 18000, 25000]
num_stores_A = 2
num_stores_B = 3

optimal_locations = find_optimal_locations(population_distribution, num_stores_A, num_stores_B)
print("Optimal Locations for Chain A:", optimal_locations[0])
print("Optimal Locations for Chain B:", optimal_locations[1])
```

#### Solution: 
1. To represent researchers and institutions, I would create Python classes `Researcher` and `Institution`. Each `Researcher` object would have attributes like `name`, `areas_of_expertise`, and `preferred_topics`. Each `Institution` object would have attributes like `name`, `desired_research_areas`, and `capacity_for_projects`.

   Example:

```python
class Researcher:
    def __init__(self, name, areas_of_expertise, preferred_topics):
        self.name = name
        self.areas_of_expertise = areas_of_expertise
        self.preferred_topics = preferred_topics

class Institution:
    def __init__(self, name, desired_research_areas, capacity_for_projects):
        self.name = name
        self.desired_research_areas = desired_research_areas
        self.capacity_for_projects = capacity_for_projects
```

2. The algorithm would iterate through each researcher and find the institution that best matches their expertise and preferred topics. Similarly, it would iterate through each institution and allocate the researcher that aligns most closely with their desired research areas.  

This algorithm ensures that both researchers and institutions are mutually satisfied with the collaborations, as it maximizes compatibility.  

Here is a Python function implementing this algorithm:

```python
def match_researchers_to_institutions(researchers, institutions):
    matching = {}
    for researcher in researchers:
        best_match = None
        max_similarity = -1
        for institution in institutions:
            similarity = calculate_similarity(researcher, institution)
            if similarity > max_similarity and institution not in matching.values():
                best_match = institution
                max_similarity = similarity
        matching[researcher] = best_match
    return matching
```

4. I will provide a sample set of researchers and institutions along with their preferences and run the `match_researchers_to_institutions` function to demonstrate the allocation process. Below is the complete Python code with sample data for the simulation:

```python
class Researcher:
    def __init__(self, name, areas_of_expertise, preferred_topics):
        self.name = name
        self.areas_of_expertise = areas_of_expertise
        self.preferred_topics = preferred_topics

class Institution:
    def __init__(self, name, desired_research_areas, capacity_for_projects):
        self.name = name
        self.desired_research_areas = desired_research_areas
        self.capacity_for_projects = capacity_for_projects

def calculate_similarity(researcher, institution):
    expertise_intersection = len(set(researcher.areas_of_expertise) & set(institution.desired_research_areas))
    return expertise_intersection

def match_researchers_to_institutions(researchers, institutions):
    matching = {}
    for researcher in researchers:
        best_match = None
        max_similarity = -1
        for institution in institutions:
            similarity = calculate_similarity(researcher, institution)
            if similarity > max_similarity and institution not in matching.values():
                best_match = institution
                max_similarity = similarity
        matching[researcher] = best_match
    return matching

# Sample Data
researcher1 = Researcher("John Doe", ["Machine Learning", "Natural Language Processing"], ["Sentiment Analysis", "Recommendation Systems"])
researcher2 = Researcher("Jane Smith", ["Computer Vision", "Robotics"], ["Object Detection", "Autonomous Navigation"])
researcher3 = Researcher("Sam Johnson", ["Data Science", "Statistics"], ["Predictive Modeling", "Statistical Inference"])

institution1 = Institution("Tech Institute", ["Machine Learning", "Data Science"], 2)
institution2 = Institution("AI Research Center", ["Machine Learning", "Computer Vision"], 1)

researchers = [researcher1, researcher2, researcher3]
institutions = [institution1, institution2]

# Run the Matching Algorithm
matching_result = match_researchers_to_institutions(researchers, institutions)

# Print the Matching Result
for researcher, institution in matching_result.items():
    print(f"{researcher.name} is matched with {institution.name if institution else 'No Institution'}")
```

In this simulation, we have three researchers (`researcher1`, `researcher2`, and `researcher3`) with their areas of expertise and preferred topics, and two institutions (`institution1` and `institution2`) with their desired research areas and project capacities.

The `match_researchers_to_institutions` function finds the best match for each researcher based on the similarity of expertise areas.

After running the algorithm, it prints the result showing which researcher is matched with which institution. If a researcher is not matched with any institution, it will display "No Institution".
