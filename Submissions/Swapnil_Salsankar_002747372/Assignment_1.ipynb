A.	Write a Python implementation of the Gale-Shapley algorithm to match residents to hospitals. Assume there are 10 hospitals and 30 residents, with each hospital having 3 slots available. You can make up preference lists for the residents and hospitals.

```python
 import random

# Define the number of residents and hospitals
n_residents = 30
n_hospitals = 10

# Define the number of slots available for each hospital
slots_per_hospital = 3

# Generate preference lists for residents and hospitals
# Residents rank hospitals, and hospitals rank residents
# Residents are represented by integers 0 to n_residents - 1
# Hospitals are represented by integers n_residents to n_residents + n_hospitals - 1

resident_prefs = {}
hospital_prefs = {}

for resident in range(n_residents):
    # Generate a random preference list of hospitals for this resident
    hospital_list = list(range(n_residents, n_residents + n_hospitals))
    random.shuffle(hospital_list)
    resident_prefs[resident] = hospital_list

for hospital in range(n_hospitals):
    # Generate a random preference list of residents for this hospital
    resident_list = list(range(n_residents))
    random.shuffle(resident_list)
    hospital_prefs[hospital + n_residents] = resident_list

# Implement the matching algorithm
matches = {}
unmatched_residents = list(range(n_residents))

while unmatched_residents:
    resident = unmatched_residents.pop(0)
    resident_prefs_list = resident_prefs[resident]

    for hospital in resident_prefs_list:
        if hospital not in matches:
            matches[hospital] = [resident]
            break
        elif len(matches[hospital]) < slots_per_hospital:
            matches[hospital].append(resident)
            break
        else:
            current_residents = matches[hospital]
            hospital_prefs_list = hospital_prefs[hospital]
            if hospital_prefs_list.index(resident) < hospital_prefs_list.index(current_residents[-1]):
                unmatched_residents.append(current_residents.pop())
                matches[hospital].append(resident)
                break

# Print the matches
for hospital, residents in matches.items():
    print(f"Hospital {hospital} matched with residents {residents}")
```

B.	Consider a variation of the problem where there are two rounds of matching. In the first round, residents apply to hospitals, and in the second round, hospitals apply to residents. Modify the Gale-Shapley algorithm to account for this and write a Python implementation.

```python
def two_round_gale_shapley(res_pref, hos_pref):

# Round 1: residents propose to hospitals
    hos_proposals = {}
    for res in res_pref:
        for hos in res_pref[res]:
            if hos not in hos_proposals:
                hos_proposals[hos] = []
            hos_proposals[hos].append(res)
            break

    res_matches = {}
    hos_matches = {}
    while hos_proposals:
        hos, res_list = hos_proposals.popitem()
        for res in res_list:
            if res not in res_matches:
                res_matches[res] = hos
                hos_matches[hos] = res
                break
            else:
                curr_hos = res_matches[res]
                if hos_pref[res].index(hos) < hos_pref[res].index(curr_hos):
                    res_matches[res] = hos
                    hos_matches[hos] = res
                    del hos_matches[curr_hos]
                    hos_proposals[curr_hos].extend([r for r in res_pref if r != res])
                    break

    # Round 2: hospitals propose to residents
    res_proposals = {}
    for hos in hos_pref:
        for res in hos_pref[hos]:
            if res not in res_matches:
                if res not in res_proposals:
                    res_proposals[res] = []
                res_proposals[res].append(hos)
                break

    while res_proposals:
        res, hos_list = res_proposals.popitem()
        for hos in hos_list:
            if hos not in hos_matches:
                res_matches[res] = hos
                hos_matches[hos] = res
                break
            else:
                curr_res = hos_matches[hos]
                if res_pref[hos].index(res) < res_pref[hos].index(curr_res):
                    res_matches[res] = hos
                    hos_matches[hos] = res
                    del res_matches[curr_res]
                    res_proposals[curr_res].extend([h for h in hos_pref if h != hos])
                    break

    return res_matches

res_pref = {
    'R1': ['H2', 'H3', 'H1'],
    'R2': ['H3', 'H1', 'H2'],
    'R3': ['H1', 'H2', 'H3']
}

hos_pref = {
    'H1': ['R2', 'R1', 'R3'],
    'H2': ['R1', 'R3', 'R2'],
    'H3': ['R1', 'R2', 'R3']
}

res_matches = two_round_gale_shapley(res_pref, hos_pref)
print(res_matches)
```

```python
from sympy import symbols, simplify
from sympy.calculus.util import function_range
import sympy

def compare_algorithm_efficiency(f_expression, g_expression):
    # Define the variable 'n' as a symbol
    n = symbols('n')

    # Parse the input expressions and simplify them
    f = simplify(sympy.sympify(f_expression))
    g = simplify(sympy.sympify(g_expression))

    # Calculate the limits of the functions as n approaches infinity
    limit_f = sympy.limit(f, n, sympy.oo)
    limit_g = sympy.limit(g, n, sympy.oo)

    # Compare the limits to determine efficiency
    if limit_f > limit_g:
        return "Algorithm A is more efficient."
    elif limit_f < limit_g:
        return "Algorithm B is more efficient."
    else:
        return "Both algorithms have the same efficiency."

# Example usage:
if __name__ == "__main__":
    f_expression = "3*n**2 + 2*n + 1"
    g_expression = "2*n**3 - 4*n**2"

    result = compare_algorithm_efficiency(f_expression, g_expression)
    print(result)
```

**Problem Statement:**

Identifying Articulation Points and Bridges

You are given an undirected graph, and your task is to identify the articulation points (nodes whose removal disconnects the graph) and bridges (edges whose removal increases the number of connected components) using Depth-First Search (DFS).

**Input Format:**

```python
from collections import defaultdict

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = defaultdict(list)
        self.time = 0

    def add_edge(self, u, v):
        self.graph[u].append(v)
        self.graph[v].append(u)

    def _dfs(self, u, visited, parent, disc, low, ap):
        children = 0
        disc[u] = self.time
        low[u] = self.time
        self.time += 1
        visited[u] = True

        for v in self.graph[u]:
            if not visited[v]:
                children += 1
                self._dfs(v, visited, u, disc, low, ap)
                low[u] = min(low[u], low[v])

                # Check for articulation point
                if parent == -1 and children > 1:
                    ap[u] = True
                if parent != -1 and low[v] >= disc[u]:
                    ap[u] = True

                # Check for bridges
                if low[v] > disc[u]:
                    print(f"Bridge: ({u}, {v})")

            elif v != parent:
                low[u] = min(low[u], disc[v])

    def find_articulation_points_and_bridges(self):
        visited = [False] * self.V
        disc = [float("inf")] * self.V
        low = [float("inf")] * self.V
        ap = [False] * self.V

        for i in range(self.V):
            if not visited[i]:
                self._dfs(i, visited, -1, disc, low, ap)

        print("Articulation Points:")
        for i, is_ap in enumerate(ap):
            if is_ap:
                print(i)

# Example usage:
# g = Graph(5)
# g.add_edge(0, 1)
# g.add_edge(0, 2)
# g.add_edge(1, 2)
# g.add_edge(2, 3)
# g.add_edge(3, 4)

g = Graph(7)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(3, 4)
g.add_edge(3, 5)
g.add_edge(6, 3)

print("Articulation Points and Bridges:")
g.find_articulation_points_and_bridges()
```

The Word Ladder Problem is a classic problem in artificial intelligence and graph theory. The goal is to find the shortest transformation sequence from a start word to an end word by changing one letter at a time, with each intermediate step being a valid English word. Heuristic search algorithms like A* can be used to efficiently solve this problem. Here's how you can approach it:

**Problem Statement:** Word Ladder Problem

Given a start word, an end word, and a list of valid English words, find the shortest transformation sequence from the start word to the end word. Each transformation step involves changing one letter of the current word to create a new valid word. The sequence must consist of valid words, and each intermediate step must be one letter different from the previous word. If no such sequence exists, return an empty sequence.

**Input Format:**

Two strings, startWord and endWord, representing the start and end words.

A list of strings, wordList, representing valid English words.

The input words and words in the word list consist of lowercase English letters only.

**Output Format:**

Sample Input and Output:

```python
from collections import defaultdict, deque

def findLadders(beginWord, endWord, wordList):
    # Create a set of valid words for faster lookup
    wordSet = set(wordList)

    # Handle the case where the endWord is not in the wordList
    if endWord not in wordSet:
        return []

    # Create a dictionary to store neighbors of words
    neighbors = defaultdict(list)
    for word in wordList:
        for i in range(len(word)):
            pattern = word[:i] + '*' + word[i+1:]
            neighbors[pattern].append(word)

    # Initialize the queue for BFS
    queue = deque([(beginWord, [beginWord])])
    visited = set()
    result = []
    found = False

    while queue:
        current, path = queue.popleft()
        visited.add(current)

        for i in range(len(current)):
            pattern = current[:i] + '*' + current[i+1:]
            for neighbor in neighbors[pattern]:
                if neighbor == endWord:
                    found = True
                    result.append(path + [neighbor])
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor]))

        if found:
            break

    return result

# Example usage:
startWord = "hit"
endWord = "cog"
wordList = ["hot", "dot", "dog", "lot", "log", "cog"]
result = findLadders(startWord, endWord, wordList)
print(result)  # Output: [['hit', 'hot', 'dot', 'dog', 'cog']]
```
