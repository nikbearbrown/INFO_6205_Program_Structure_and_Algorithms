Arrange the following functions in increasing order of growth: 
1. $2^n$
2. $log_2(n^2)$

5. $\log_2(\log_2(n))$
6. $n^3$
7. $3^n$
8. $10n$
9. $(\log_2(n))^2$
10. $n!$

<font color=red>Solution:</font>

1. $2^n$:
   Exponential growth, which is rapid. For our set, this is outpaced only by the factorial function.

2. $log_2(n^2)$:
   Using logarithm properties, this becomes $2 \log_2(n)$. This is still a logarithmic function, so it has a growth rate of $O(\log(n))$.

3. $5n^2 + 4n$:
   A polynomial function where the dominant term is $n^2$. So, its growth rate is $O(n^2)$.

4. $n \cdot log_2(n)$:
   This is a logarithmic-linear function, or commonly known as $O(n \log n)$, which grows faster than linear functions but slower than quadratic functions.

5. $log_2(\log_2(n))$:
   This is even slower-growing than a regular logarithm.

6. $n^3$:
   A cubic polynomial, so its growth rate is $O(n^3)$.

7. $3^n$:
   Another exponential function but with a base larger than the first. This grows faster than $2^n$ but slower than $n!$.

8. $10n$:
   A linear function with a growth rate of $O(n)$.

9. $(\log_2(n))^2$:
   Logarithm squared, meaning it grows faster than a simple logarithm but much slower than any polynomial of degree 1 or higher.

10. $n!$:
   Factorial growth, which outpaces all the other functions in the list for sufficiently large $n$.

Given the analysis, the functions in increasing order of growth are:

1. $log_2(\log_2(n))$
2. $log_2(n^2) $
3. $(\log_2(n))^2$
4. $10n$
5. $n \cdot log_2(n)$
6. $5n^2 + 4n$
7. $n^3$
8. $2^n$
9. $3^n$
10. $n!$

So, as $n$ goes to infinity, $n!$ grows the fastest while $log_2(\log_2(n))$ grows the slowest.

1. $h(2n) = O(k(n))$
2. $h(n^2) = O(k(2n))$
3. If $h(n) = O(k(n))$, does it imply $k(n^3) = \Omega(h(n^2))$?
4. $h(n) + k(n) = O(h(n) \cdot k(n))$
5. $h(n) \cdot k(n) = O(h(n) + k(n))$

Evaluate the truthiness of each of the above statements and provide reasoning or examples.


<font color=red>Solution:</font>
1. $h(2n) = O(k(n))$

**Claim:** False.

**Counterexample:** 
Consider $h(n) = n$ and $k(n) = n$. Clearly, $h(n) = O(k(n))$. However, $h(2n) = 2n$, which is not $O(n)$. Therefore, the claim is false.

2. $h(n^2) = O(k(2n))$

**Claim:** False.

**Counterexample:** 
Consider $h(n) = 2^n$ and $k(n) = n$. While $h(n) = O(2^n)$, $h(n^2) = 2^{n^2}$ clearly grows much faster than $k(2n) = 2n$, and thus cannot be bounded by it. Hence, the claim is false.

3. If $h(n) = O(k(n))$, does it imply $k(n^3) = \Omega(h(n^2))$?

**Claim:** False.

**Counterexample:** 
Let's use $h(n) = n^2$ and $k(n) = n^3$. While $h(n) = O(k(n))$, $k(n^3) = n^9$, which is not bounded below by a constant multiple of $h(n^2) = n^4$. Therefore, the statement is false.

4. $h(n) + k(n) = O(h(n) \cdot k(n))$

**Claim:** True.

**Proof:** 
Given $h(n) = O(k(n))$, there exist constants $c$ and $n_0$ such that $0 \leq h(n) \leq c \cdot k(n)$ for all $n > n_0$. 
Now, $h(n) + k(n) \leq c \cdot k(n) + k(n) = (c+1) \cdot k(n)$. Clearly, $h(n) \cdot k(n)$ will grow faster than this, so the claim is true. 

5. $h(n) \cdot k(n) = O(h(n) + k(n))$

**Claim:** False.

**Counterexample:** 
Consider $h(n) = n$ and $k(n) = n$. Then $h(n) \cdot k(n) = n^2$, while $h(n) + k(n) = 2n$. Clearly, $n^2$ cannot be bounded by a constant multiple of $2n$. Therefore, the statement is false.

These solutions provide a basis for understanding the relationships between these functions in big O notation and showcase the importance of choosing functions carefully when evaluating growth rates.

Suppose in a university, there are $ y $ distinct research projects, each having a specified number of positions for graduate students. There are $ m $ graduate students, each looking to join a research project. We will assume that there are more research projects than the total positions available across all projects.

Each research project has a ranking for students based on their prior work, academic achievements, and research interests. Similarly, each student has a ranking for the research projects based on the project's relevance to their interests, the prominence of the principal investigator, and potential career benefits.

The goal is to assign each student to at most one research project, ensuring all positions in a project are filled. There might be some projects without any students due to the constraints.

A pairing is considered stable if:

1. No student $ s_1 $ is matched to a project $ p $ while another student $ s_2 $ is unmatched, and project $ p $ prefers $ s_2 $ over $ s_1 $.
2. No students $ s_1 $ and $ s_2 $ are matched to projects $ p_1 $ and $ p_2 $ respectively, such that:
   - $ s_1 $ prefers $ p_2 $ over $ p_1 $, and
   - $ p_2 $ prefers $ s_1 $ over $ s_2 $.

Under these circumstances:

1. Can you prove that a stable assignment of students to research projects always exists?
2. Propose an algorithm to find one such stable assignment.

<font color=red>Solution:</font>   
    

**Proof of Existence:**

We can show the existence of a stable assignment by demonstrating that the Gale-Shapley algorithm will always terminate and will result in a stable assignment:

1. Every iteration of the algorithm reduces the number of unmatched students (unless they've proposed to all projects).
2. The algorithm ends when every student is matched or has proposed to every project.
3. By the end, no project has an unfilled slot because, before moving on to a less preferred student, every student will propose to every project.

**Gale-Shapley Algorithm for University Research Projects:**

1. Initially, all students are unmatched and all research projects have their full capacities available.
2. Each student proposes to their top-choice research project.
3. Research projects, on receiving proposals:
   - If the project hasn't reached its capacity, it temporarily accepts the student.
   - If the project has reached its capacity but prefers the new student over one it has already accepted, it rejects the least-preferred student and temporarily accepts the new student. The rejected student becomes unmatched.
   - If the project has reached its capacity and does not prefer the new student over any of its current students, it rejects the new student.
4. Unmatched students propose to their next-choice projects.
5. Continue the process until every student has either been accepted by a project or has been rejected by all projects.

**Proof of Stability:**

- For any student $ s_1 $ matched to project $ p_1 $, there cannot be another student $ s_2 $ not matched to any project that $ p_1 $ prefers over $ s_1 $, because if that were the case, $ p_1 $ would have accepted $s_2 $ before $ s_1 $.
  
- For any student $ s_1 $ matched to project $ p_1 $ and any student $ s_2 $ matched to project $ p_2 $, where $ s_2 $ prefers $ p_1 $ over $ p_2 $ and $ p_1 $ prefers $ s_2 $ over $ s_1 $, this situation is impossible. This is because $ s_2 $ would have proposed to $ p_1 $ before proposing to $ p_2 $. If $ p_1$ preferred $ s_2 $ over its least-preferred student at that time, it would have accepted $s_2 $ and not $ s_1$, contradicting our assumption.

This ensures that the final matching is stable.

A tower is constructed by stacking cylindrical drums. The total height of drums in each level is the summation of the height of drums in the current level and the level below. We are in the process of constructing the tower with these drums. An asymptotic analysis can be applied here.

Assume that:
1. Each level of the tower can hold drums with a combined height of at most a constant $d$.
2. The entire tower will be constructed with drums having a total height of at most $h$.

Given the constraints, how can you design the tower such that the drums have a height $g(h)$, where $g(h)$ is a function that grows as slowly as possible? Describe the function $g(h)$ and explain the construction process.

<font color=red> Solution:</font>   

1. **Tower Construction**:
    - The tower is built level by level.
    - If we denote the height of drums at level $i$ as $D_i$, then the relationship is $D_i = D_{i-1} + D_{i-2}$.
    - The beginning levels (initial conditions) can be set as $D_1 = 1$ and $D_2 = 1$, based on the Fibonacci sequence logic.

2. **Space Requirement**:
    - For each level, the height occupied by the drums is limited by the constant $d$.
    - The total height occupied by all levels together cannot exceed $h$.
    - The height occupied by all the drums for the tower is $D_1 + D_2 + ... + D_R \leq h$, where $R$ is the number of levels.

3. **Relating Total Height to Levels**:
    - Given the relationship based on the Fibonacci-like sequence, the total height can be expressed as the sum of the first $R$ elements of a modified Fibonacci sequence, which is $\frac{R(R+1)}{2}$. 
    - From this, we get the relationship: $\frac{1}{2} R(R+1) \leq h$.
    - Solving for $R$, we deduce: $R \leq \sqrt{2h}$.

4. **Function $g(h)$**:
    - Using the relationship derived above, the height of drums at any level based on the maximum height $h$ is given by: $g(h) \leq d + d\sqrt{2h}$.
    - Therefore, the height function $g(h)$ grows as $O(\sqrt{h})$.

### Construction Process:

1. Begin with drums of height 1 for the first level.
2. Use drums of height 1 for the second level.
3. For each subsequent level, stack drums such that their combined height is the sum of the heights from the two previous levels, ensuring it doesn’t exceed $d$ or the total $h$.
4. Continue this process until the combined height of all drums reaches $h$.

A strategy pair $ (E, N) $ is deemed stable if no team from either league can adjust its strategy unilaterally to win more games. This implies there's no strategy $ E' $ such that a EuroLeague team wins more games with $ (E', N) $, and likewise, there's no strategy $ N' $ such that an NBA team wins more games with $ (E, N') $.

Given the players and their ratings for each role:

B. Implement a loop to mix the preference lists for each team 1000 times. Determine the proportion of stable group stage matchups using these shuffled lists. Consider utilizing the function random.shuffle(x[, random]).


<font color=red> Solution:</font>    

1. Create a function to check if a matching is stable.
2. Shuffle the preference lists 1000 times and use the Gale-Shapley algorithm to find matchings.
3. For each matching, check if it's stable. If it is, increment the stable match count.
4. Calculate the percentage of stable playoff matches.    
Here is code: 

```python
import random

# Gale-Shapley Matching
def gale_shapley(men_prefs, women_prefs):
    free_men = list(men_prefs.keys())
    matchings = {}
    
    while free_men:
        man = free_men.pop(0)
        man_list = men_prefs[man]
        
        for woman in man_list:
            fiance = matchings.get(woman)
            if not fiance:
                matchings[woman] = man
                break

            woman_list = women_prefs[woman]
            if woman_list.index(man) < woman_list.index(fiance):
                free_men.append(fiance)
                break
                
    return matchings

# Initial teams and preferences
initial_teams_epl = ['EPL'+str(i) for i in range(1,9)]
initial_teams_la_liga = ['LaLiga'+str(i) for i in range(1,9)]
initial_prefs_epl = {team: initial_teams_la_liga.copy() for team in initial_teams_epl}
initial_prefs_la_liga = {team: initial_teams_epl.copy() for team in initial_teams_la_liga}
for team in initial_prefs_epl:
    random.shuffle(initial_prefs_epl[team])
for team in initial_prefs_la_liga:
    random.shuffle(initial_prefs_la_liga[team])

# Get an initial matching to check stability later
initial_matching = gale_shapley(initial_prefs_epl, initial_prefs_la_liga)

stable_match_count = 0
iterations = 1000

for _ in range(iterations):
    for team in initial_prefs_epl:
        random.shuffle(initial_prefs_epl[team])
    for team in initial_prefs_la_liga:
        random.shuffle(initial_prefs_la_liga[team])

    new_matching = gale_shapley(initial_prefs_epl, initial_prefs_la_liga)
    
    if new_matching == initial_matching:
        stable_match_count += 1

percentage_stable_matches = (stable_match_count / iterations) * 100
print(f"Percentage of Stable Matchups: {percentage_stable_matches}%")
```

C. Based on fictional match outcomes, determine which teams proceed and which get eliminated in each round. Following this, remove the ousted teams from the preference lists for the upcoming rounds. Can you continuously apply the Gale-Shapley matching algorithm for each phase (32 teams, 16 teams, 8 teams, 4 teams) to achieve stable draws?


```python
import random

def gale_shapley(epl, la_liga, epl_prefs, la_liga_prefs):
    free_teams = epl.copy()
    matches = {}
    proposals = {team: [] for team in epl}
    
    while free_teams:
        team_epl = free_teams.pop()
        for team_la_liga in epl_prefs[team_epl]:
            if team_la_liga not in proposals[team_epl]:
                proposals[team_epl].append(team_la_liga)
                if team_la_liga not in matches:
                    matches[team_la_liga] = team_epl
                    break
                elif la_liga_prefs[team_la_liga].index(team_epl) < la_liga_prefs[team_la_liga].index(matches[team_la_liga]):
                    free_teams.append(matches[team_la_liga])
                    matches[team_la_liga] = team_epl
                    break
    return matches

def eliminate_teams(matches):
    # This function randomly eliminates one team from each match
    return random.choice([team for pair in matches.items() for team in pair])

# Start the playoff with the initial teams and preferences
epl_remaining = epl.copy()
la_liga_remaining = la_liga.copy()
epl_prefs_remaining = epl_prefs.copy()
la_liga_prefs_remaining = la_liga_prefs.copy()

rounds = [16, 8, 4, 2]

for r in rounds:
    print(f"Round with {r} teams:")
    matches = gale_shapley(epl_remaining, la_liga_remaining, epl_prefs_remaining, la_liga_prefs_remaining)
    print(matches)
    
    # Eliminate half of the teams
    eliminated = [eliminate_teams(matches) for _ in range(r // 2)]
    print(f"Eliminated: {eliminated}")
    
    # Update remaining teams and their preferences
    epl_remaining = [team for team in epl_remaining if team not in eliminated]
    la_liga_remaining = [team for team in la_liga_remaining if team not in eliminated]
    epl_prefs_remaining = {team: prefs for team, prefs in epl_prefs_remaining.items() if team not in eliminated}
    la_liga_prefs_remaining = {team: prefs for team, prefs in la_liga_prefs_remaining.items() if team not in eliminated}
    for team, prefs in epl_prefs_remaining.items():
        epl_prefs_remaining[team] = [p for p in prefs if p not in eliminated]
    for team, prefs in la_liga_prefs_remaining.items():
        la_liga_prefs_remaining[team] = [p for p in prefs if p not in eliminated]

print("Playoff complete!")
```

D. Merge the team lists, disregarding their leagues, so that any team can potentially face any other. In this scenario, does the Gale-Shapley matching algorithm still establish stable matchups, especially when it's a single list matching within itself?


```python
def gale_shapley_all_teams(all_teams, prefs):
    free_teams = all_teams.copy()
    matches = {}
    proposals = {team: [] for team in all_teams}
    
    while free_teams:
        team1 = free_teams.pop()
        for team2 in prefs[team1]:
            if team2 != team1 and team2 not in proposals[team1]:  # Team shouldn't match with itself
                proposals[team1].append(team2)
                if team2 not in matches:
                    matches[team2] = team1
                    break
                elif prefs[team2].index(team1) < prefs[team2].index(matches[team2]):
                    free_teams.append(matches[team2])
                    matches[team2] = team1
                    break
                    
    return matches

# Combining the teams from both conferences
all_teams = epl + la_liga

# Combining the preferences
all_prefs = {}
for team in epl:
    all_prefs[team] = [t for t in all_teams if t != team]  # Excluding the team itself from preferences

for team in la_liga:
    all_prefs[team] = [t for t in all_teams if t != team]  # Excluding the team itself from preferences

matches = gale_shapley_all_teams(all_teams, all_prefs)
print(matches)
```

```python
import time
import random

# Gale-Shapley Matching
def gale_shapley(men_prefs, women_prefs):
    free_men = list(men_prefs.keys())
    matchings = {}
    
    while free_men:
        man = free_men.pop(0)
        man_list = men_prefs[man]
        
        for woman in man_list:
            fiance = matchings.get(woman)
            if not fiance:
                matchings[woman] = man
                break

            woman_list = women_prefs[woman]
            if woman_list.index(man) < woman_list.index(fiance):
                free_men.append(fiance)
                break
                
    return matchings

# Function to double the teams and preferences
def double_teams_and_prefs(original_teams):
    new_teams = original_teams + ["FC" + chr(i) for i in range(65, 65 + len(original_teams))]
    new_prefs = {team: new_teams.copy() for team in new_teams}
    for team, prefs in new_prefs.items():
        random.shuffle(prefs)
    return new_teams, new_prefs

# Measure Gale-Shapley execution time
def measure_time_for_gale_shapley(men_prefs, women_prefs):
    start_time = time.time()
    gale_shapley(men_prefs, women_prefs)
    return time.time() - start_time

# Initial teams
initial_teams_epl = ['EPL'+str(i) for i in range(1,9)]
initial_teams_la_liga = ['LaLiga'+str(i) for i in range(1,9)]
initial_prefs_epl = {team: initial_teams_la_liga.copy() for team in initial_teams_epl}
initial_prefs_la_liga = {team: initial_teams_epl.copy() for team in initial_teams_la_liga}
for team in initial_prefs_epl:
    random.shuffle(initial_prefs_epl[team])
for team in initial_prefs_la_liga:
    random.shuffle(initial_prefs_la_liga[team])

team_counts = [len(initial_teams_epl)]
times = [measure_time_for_gale_shapley(initial_prefs_epl, initial_prefs_la_liga)]

# Double the size and measure execution time
for _ in range(5):  # Doubling 5 times as an example
    initial_teams_epl, initial_prefs_epl = double_teams_and_prefs(initial_teams_epl)
    initial_teams_la_liga, initial_prefs_la_liga = double_teams_and_prefs(initial_teams_la_liga)
    elapsed_time = measure_time_for_gale_shapley(initial_prefs_epl, initial_prefs_la_liga)
    times.append(elapsed_time)
    team_counts.append(len(initial_teams_epl))

print("Number of teams:", team_counts)
print("Execution times:", times)
```
