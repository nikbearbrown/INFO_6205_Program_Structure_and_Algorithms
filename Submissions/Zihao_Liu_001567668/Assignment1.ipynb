## Q2:
Consider two sets, $ A = \{a_1, a_2\} $ and $ B = \{b_1, b_2\}$, where $ A $ and $ B $ are two different sets of individuals seeking to be matched with each other. The preference lists for the individuals are as follows:

- $ a_1 $: $ b_1, b_2 $
- $ a_2 $: $ b_2, b_1 $
- $ b_1 $: $ a_2, a_1 $
- $ b_2 $: $ a_1, a_2 $

Apply the Gale-Shapley Algorithm to find a stable matching. Is there more than one stable matching for the given preference lists?

### Answer 2:
Let each individual in A propose to their preferred individual in set B. Base on the preference list of B. each individual in set B either accepting the proposal or rejecting it.

1. $ a_1 $ proposes to $ b_1 $ and $ a_2 $ proposes to $ b_2 $.
2. $ b_1 $ prefers $ a_2 $ to $ a_1 $ and $ b_2 $ prefers $ a_1 $ to $ a_2 $.
3. So, $ b_1 $ rejects $ a_1 $ and $ b_2 $ rejects $ a_2 $.
4. Now, $ a_1 $ proposes to $ b_2 $ and $ a_2 $ proposes to $ b_1 $.

Thus rhe stable matching is $ \{(a_1, b_2), (a_2, b_1)\} $.

## Q3:

### Problem Statement

**Objective:**

Create a program that takes two lists, one of applicants and one of companies, where each applicant has a list of preferred companies and each company has a list of preferred applicants. The program should output stable matches.

**Background:**

A matching is stable if there exists no applicant-company pair (A, C) such that:
- A is matched with C' but prefers C over C'.
- C is matched with A' but prefers A over A'.

### Inputs:

- A list `applicants`, where each applicant is represented as a dictionary with the following keys:
  - `'name'`: a string representing the name of the applicant.
  - `'preferences'`: a list of strings representing the names of companies in decreasing order of preference.
  
- A list `companies`, where each company is represented as a dictionary with the following keys:
  - `'name'`: a string representing the name of the company.
  - `'preferences'`: a list of strings representing the names of applicants in decreasing order of preference.

### Output:

- A list of dictionaries representing the stable matches, where each dictionary has the following keys:
  - `'applicant'`: a string representing the name of the applicant.
  - `'company'`: a string representing the name of the company.

### Constraints:

- 1 ≤ |applicants|, |companies| ≤ 1000
- All names of applicants and companies are unique and consist of lowercase and uppercase English letters, and have length at most 100.
- Each applicant has a list of all companies in their preferences list, and each company has a list of all applicants in their preferences list.

### Example:

#### Input:

```python
applicants = [
    {'name': 'Alice', 'preferences': ['Google', 'Facebook']},
    {'name': 'Bob', 'preferences': ['Facebook', 'Google']}
]

companies = [
    {'name': 'Google', 'preferences': ['Alice', 'Bob']},
    {'name': 'Facebook', 'preferences': ['Bob', 'Alice']}
]
```

#### Output:

```python
[
    {'applicant': 'Alice', 'company': 'Google'},
    {'applicant': 'Bob', 'company': 'Facebook'}
]
```

### Explanation:

In this example:
- Alice prefers to work at Google over Facebook, and Google prefers Alice over Bob, so they form a stable match.
- Bob prefers to work at Facebook over Google, and Facebook prefers Bob over Alice, so they form a stable match.

````python
def stable_matching(applicants, companies):
    # Initialize the company and applicant dictionaries with their preferences
    applicant_pref_dict = {a['name']: a['preferences'] for a in applicants}
    company_pref_dict = {c['name']: c['preferences'] for c in companies}
    
    # Dictionary to keep track of the applicants' current companies
    applicant_current_company = {a['name']: None for a in applicants}
    
    # Dictionary to keep track of companies' current applicants
    company_current_applicants = {c['name']: [] for c in companies}
    
    # List of free applicants
    free_applicants = [a['name'] for a in applicants]
    
    while free_applicants:
        applicant = free_applicants.pop(0)
        prefs = applicant_pref_dict[applicant]
        
        # Iterate through the applicant's preferences
        for company in prefs:
            # If applicant is not matched with any company
            if applicant_current_company[applicant] is None:
                # Check the company's preferences
                company_prefs = company_pref_dict[company]
                
                # If company has no applicants yet
                if not company_current_applicants[company]:
                    # Match the applicant and company
                    applicant_current_company[applicant] = company
                    company_current_applicants[company].append(applicant)
                    break
                else:
                    # If company prefers the new applicant over the current one
                    current_applicant = company_current_applicants[company][0]
                    if company_prefs.index(applicant) < company_prefs.index(current_applicant):
                        # Set the current applicant as free
                        applicant_current_company[current_applicant] = None
                        free_applicants.append(current_applicant)
                        
                        # Match the new applicant and company
                        applicant_current_company[applicant] = company
                        company_current_applicants[company][0] = applicant
                        break
                    
    # Formulate the output
    result = []
    for applicant, company in applicant_current_company.items():
        result.append({'applicant': applicant, 'company': company})
        
    return result


# Example usage:
applicants = [
    {'name': 'Alice', 'preferences': ['Google', 'Facebook']},
    {'name': 'Bob', 'preferences': ['Facebook', 'Google']}
]

companies = [
    {'name': 'Google', 'preferences': ['Alice', 'Bob']},
    {'name': 'Facebook', 'preferences': ['Bob', 'Alice']}
]

print(stable_matching(applicants, companies))
````

### Q5:
Consider a directed graph G = (V, E) where V is a set of vertices and E is a set of edges. The graph has n vertices numbered from 1 to n. You have been given a task to implement a Depth-First Search (DFS) to traverse the graph and print the vertices in the order they are visited.

You are given the following information:
- The graph may contain one or more connected components.
- Vertices are numbered, and the graph is represented using an adjacency list.
- It is guaranteed that there are no self-loops, i.e., there will be no edges from a vertex to itself.
- The graph may contain cycles.

Write a Python function named `dfs_traversal` that takes an adjacency list representing the graph and a starting vertex as input and returns a list containing the vertices in the order they are visited by a Depth-First Search (DFS). If the graph has more than one connected component, your function should still return the vertices of all components in the order they are visited.

### Function Signature:
```python
def dfs_traversal(adj_list: Dict[int, List[int]], start_vertex: int) -> List[int]:
```

### Input:
- `adj_list` (1 <= len(adj_list) <= 1000) is a dictionary where the key is a vertex (int) and the value is a list of integers representing the adjacent vertices.
- `start_vertex` (1 <= start_vertex <= n) is an integer representing the starting vertex for the DFS.

### Output:
- A list of integers representing the vertices in the order they are visited by DFS.

### Example:
#### Input:
```python
adj_list = {
    1: [2],
    2: [3, 4],
    3: [],
    4: [5],
    5: [],
    6: [7],
    7: []
}
start_vertex = 1
```

#### Output:
```python
[1, 2, 3, 4, 5, 6, 7]
```

#### Explanation:
- Starting from vertex 1, the DFS visits vertices 2, 3, and 4, in this order.
- After completing the traversal for vertex 4, it goes back and visits vertex 5.
- Vertices 6 and 7 are in a different connected component, but they are also included in the output list as they are part of the overall graph.

### Note:
You need to ensure that all vertices in the graph are visited, even if they are in different connected components. Also, take care to handle the case where cycles are present in the graph to avoid infinite loops. You can assume that the input graph is well-formed, and there are no isolated vertices (vertices not present in the adjacency list).

````python
def dfs_traversal(adj_list: dict, start_vertex: int) -> list:
    visited = set()
    result = []
    
    def dfs(vertex):
        # If the vertex is visited, then return
        if vertex in visited:
            return
        # Mark the vertex as visited
        visited.add(vertex)
        # Add the vertex to the result list
        result.append(vertex)
        # Visit all the adjacent vertices
        for neighbor in adj_list.get(vertex, []):
            dfs(neighbor)
    
    # Start the DFS from the given start_vertex
    dfs(start_vertex)
    
    # Check for any unvisited vertices and traverse them as well
    for vertex in adj_list.keys():
        if vertex not in visited:
            dfs(vertex)
            
    return result

adj_list = {
    1: [2],
    2: [3, 4],
    3: [],
    4: [5],
    5: [],
    6: [7],
    7: []
}

print(dfs_traversal(adj_list, 1))
````

```
  1 - 2 - 3
```

- Option C is correct because an undirected graph can have at most n(n - 1) / 2 edges, which corresponds to a complete graph where every pair of distinct vertices is connected by a unique edge.

- Option D is incorrect because even if a graph has the same number of edges as vertices, it may not be a tree. For example, a graph with 3 vertices and 3 edges forming a cycle is not a tree.

## Q9:

You are given a list of integers, and you are to write a function that finds the two elements in the list that sum up to a specific target. You need to return these two elements. If there are multiple pairs that match the target sum, you can return any of them. If no such pair exists, return an empty list. You are required to write a solution with a time complexity better than $O(n^2)$.

### Example:

```python
nums = [1, 2, 3, 4, 5]
target = 9
# Your function should return [4,5] (or [5,4]) since 4 + 5 = 9.
```

### Constraints:

- The list can have up to $10^5$ integers.
- The list and target will be such that a solution will always exist.

````python
def two_sum(nums, target):
    seen = set()
    for num in nums:
        if target - num in seen:
            return [num, target - num]
        seen.add(num)
    return []

print(two_sum([1, 2, 3, 4, 5], 9))
````
