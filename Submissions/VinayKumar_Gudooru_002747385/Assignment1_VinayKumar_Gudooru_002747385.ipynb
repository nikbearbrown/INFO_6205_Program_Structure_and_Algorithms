## Problem 3 (10 Points)

In a university, there are n students and n scholarship programs available. Each student has a ranked list of scholarship programs they are interested in, and each scholarship program has a ranked list of students they wish to award scholarships to. The goal is to allocate scholarships to students in a way that maximizes the students' preferences and ensures that no student or scholarship program would prefer each other over their current assignments.

Your task is to implement the Gale-Shapley algorithm to find a stable allocation of scholarships to students based on their preferences.

**Input:**

Lists of ranked preferences for n students and n scholarship programs. Each list contains the names or identifiers of their preferred scholarships or students, ranked from most preferred to least preferred.

**Output:**

A stable allocation of scholarships to students that respects their preferences and ensures that no unstable assignments exist.
Example:

Suppose there are 3 students (S1, S2, S3) and 3 scholarship programs (P1, P2, P3), with their ranked preferences as follows:

**Student Preferences:**

  - S1: P2, P1, P3
  - S2: P1, P2, P3
  - S3: P1, P3, P2

**Scholarship Program Preferences:**

  - P1: S2, S1, S3
  - P2: S3, S2, S1
  - P3: S1, S3, S2

Using the Gale-Shapley algorithm, find a stable allocation of scholarships to students that respects their preferences.

## Solution:

The Gale-Shapley algorithm for scholarship allocation works as follows:

1. Initialize all students and scholarship programs as unassigned.

2. While there exists an unassigned student S:
  - Let S's highest-ranked scholarship program who has not yet been assigned to any student be P.
  - If P is unassigned, assign P to S.
  - If P is currently assigned to a different student S', then P will compare their suitors S and S' and choose the one they prefer.
  - If P prefers S' to S, nothing happens.
  - If P prefers S to S', P becomes assigned to S, and S' becomes unassigned.
3. The algorithm terminates when all students are assigned to scholarship programs.

The solution will provide a stable allocation of scholarships to students.

```python
# Number of students and scholarships
n = 5

# Manually assigned preferences for students and scholarships
student_preferences = {
    'S1': ['P3', 'P4', 'P2', 'P1', 'P5'],
    'S2': ['P2', 'P4', 'P1', 'P3', 'P5'],
    'S3': ['P4', 'P3', 'P5', 'P1', 'P2'],
    'S4': ['P5', 'P1', 'P2', 'P4', 'P3'],
    'S5': ['P1', 'P2', 'P3', 'P4', 'P5']
}

scholarship_preferences = {
    'P1': ['S4', 'S1', 'S2', 'S3', 'S5'],
    'P2': ['S1', 'S2', 'S4', 'S3', 'S5'],
    'P3': ['S4', 'S2', 'S5', 'S3', 'S1'],
    'P4': ['S2', 'S1', 'S3', 'S4', 'S5'],
    'P5': ['S1', 'S2', 'S3', 'S4', 'S5']
}

# Initialize all students and scholarships as unassigned
students = {f'S{i}': None for i in range(1, n+1)}
scholarships = {f'P{i}': None for i in range(1, n+1)}

# Gale-Shapley algorithm for scholarship allocation
while None in students.values():
    for student in students:
        if students[student] is None:
            preferred_scholarship = student_preferences[student].pop(0)
            if scholarships[preferred_scholarship] is None:
                students[student] = preferred_scholarship
                scholarships[preferred_scholarship] = student
            else:
                current_student = scholarships[preferred_scholarship]
                if scholarship_preferences[preferred_scholarship].index(student) < scholarship_preferences[preferred_scholarship].index(current_student):
                    students[student] = preferred_scholarship
                    students[current_student] = None
                    scholarships[preferred_scholarship] = student

# Print the stable allocation
for student, scholarship in students.items():
    print(f'Student {student} is assigned to Scholarship {scholarship}')
```

Let's implement one of the optimizations mentioned earlier: "Parallelization." We'll use Python's `concurrent.futures` module to parallelize the preference evaluation step. This optimization can potentially speed up the stable matching algorithm for large datasets by allowing multiple student proposals to companies to be evaluated concurrently.

Here's the modified code with parallelization:

```python
import concurrent.futures

def stable_matching(companies, students):
    # Initialize an empty matching for students and companies
    matching = {}
    
    # Define the preferences of companies and students
    company_preferences = {
        "Company A": ["S1", "S2", "S3"],
        "Company B": ["S2", "S1", "S3"],
        "Company C": ["S2", "S1", "S3"]
    }
    
    student_preferences = {
        "S1": ["Company C", "Company B", "Company A"],
        "S2": ["Company B", "Company C"],
        "S3": ["Company A"]
    }
    
    # Keep track of students who are not yet matched
    unmatched_students = set(students)
    
    # Function to evaluate student proposals to a company
    def evaluate_proposals(student, student_choices, company):
        nonlocal matching
        nonlocal unmatched_students
        
        for company in student_choices:
            if company not in matching:
                # Company is unmatched, so match the student with the company
                matching[company] = student
                break
            else:
                current_match = matching[company]
                # Check if the student is preferred over the current match
                if company_preferences[company].index(student) < company_preferences[company].index(current_match):
                    # Student is preferred, so unmatch the current match and match the student
                    matching[company] = student
                    unmatched_students.add(current_match)
                    break
    
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Evaluate student proposals concurrently
        futures = {executor.submit(evaluate_proposals, student, student_preferences[student], None): student for student in unmatched_students}
        concurrent.futures.wait(futures)
    
    return matching

# Example usage
companies = ["Company A", "Company B", "Company C"]
students = ["S1", "S2", "S3"]

matching = stable_matching(companies, students)

# Print the stable matching
for company, student in matching.items():
    print(f"{student} is matched with {company}")
```

In this modified code, we use a `ThreadPoolExecutor` to submit the evaluation of student proposals to companies concurrently. This allows multiple students to propose to different companies at the same time, potentially speeding up the algorithm for larger datasets.

## Problem 6 (10 Points)

What is the time complexity of finding the Fibonacci sequence using a recursive algorithm? Can you suggest an optimized approach to improve the time complexity?

## Solution:

The time complexity of finding the Fibonacci sequence using a simple recursive algorithm is exponential, O(2^n). This is because the algorithm repeatedly recalculates Fibonacci numbers, resulting in an exponential number of recursive calls. 

```python
def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
```

An optimized approach is to use dynamic programming or memoization, which reduces the time complexity to O(n) by storing previously computed Fibonacci numbers.

```python
def fibonacci_memoization(n, memo={}):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    elif n in memo:
        return memo[n]
    else:
        memo[n] = fibonacci_memoization(n - 1, memo) + fibonacci_memoization(n - 2, memo)
        return memo[n]
```

In this optimized version, memo is a dictionary that stores previously computed Fibonacci numbers. When a Fibonacci number is requested, the function first checks if it's in the memo dictionary. If it's already computed, it returns the stored value; otherwise, it calculates the value recursively and stores it in the memo dictionary for future use.

The time complexity of this optimized memoization approach is O(n), as it computes each Fibonacci number once and stores the result for reuse. This is a significant improvement over the exponential time complexity of the naive recursive approach.

```python
import time

# Calculate and print the first 20 Fibonacci numbers using both methods
n = 20
print("Fibonacci Numbers (Recursive):")
start_time = time.time()
for i in range(n):
    result = fibonacci_recursive(i)
    print(f"F({i}) = {result}")
end_time = time.time()
print(f"Processing Time: {end_time - start_time} seconds\n")

print("Fibonacci Numbers (Memoization):")
start_time = time.time()
memo = {}
for i in range(n):
    result = fibonacci_memoization(i, memo)
    print(f"F({i}) = {result}")
end_time = time.time()
print(f"Processing Time: {end_time - start_time} seconds")
```

## Problem 7 (10 points)

Compare the time complexities of two algorithms: one with a runtime of $n^2 + 2n$ and another with a runtime of $3n log2(n))$. Which algorithm is more efficient in terms of asymptotic time complexity? Determine the crossover point.

## Solution:

We have two algorithms with different time complexities:

Algorithm A: T(n) = n^2 + 2n

Algorithm B: T(n) = 3n log2(n)

We need to compare these algorithms in terms of asymptotic time complexity and find the crossover point where one algorithm becomes more efficient than the other.

1. **Comparing Time Complexities:**

   We are given the time complexities of Algorithm A and Algorithm B.
   - Algorithm A has a time complexity of n^2 + 2n.
   - Algorithm B has a time complexity of 3n log2(n).
   To compare the two algorithms, we want to determine which one grows faster as the input size (n) increases. In Big O notation, we are interested in the dominant term that contributes most significantly to the   growth rate.

   - For Algorithm A (n^2 + 2n), the dominant term is n^2 because it grows faster than 2n as n becomes large.
   - For Algorithm B (3n log2(n)), the dominant term is 3n log2(n) because it grows faster than 3n as n becomes large.
   
   Therefore, Algorithm B (3n log2(n)) has a better (i.e., smaller) time complexity in terms of asymptotic analysis because it grows more slowly than Algorithm A (n^2 + 2n) as the input size increases.

2. **Finding the Crossover Point:**

   To determine the crossover point, we set the two time complexities equal to each other:
   n^2 + 2n = 3n log2(n)

   The goal is to find the value of n where these two functions intersect or become equal.

```python
#Plotting graph for crossover point

import numpy as np
import matplotlib.pyplot as plt

# Define the functions for Algorithm A and Algorithm B
def algorithm_a(n):
    return n**2 + 2*n

def algorithm_b(n):
    return 3*n*np.log2(n)

# Create an array of values for n
n_values = np.arange(1, 10)  # Adjust the range as needed

# Calculate the corresponding function values
a_values = algorithm_a(n_values)
b_values = algorithm_b(n_values)

# Plot the two functions
plt.figure(figsize=(8, 6))
plt.plot(n_values, a_values, label='Algorithm A (n^2 + 2n)')
plt.plot(n_values, b_values, label='Algorithm B (3n log2(n))')

# Add labels and legend
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.legend()

# Find the crossover point (where the two functions intersect)
crossover_point = np.argwhere(np.diff(np.sign(a_values - b_values)))[0][0]
plt.plot(n_values[crossover_point], a_values[crossover_point], 'ro')  # Mark the crossover point
plt.annotate(f'Crossover Point ({n_values[crossover_point]}, {a_values[crossover_point]:.2f})', 
             (n_values[crossover_point], a_values[crossover_point]), 
             textcoords="offset points", xytext=(10,10), ha='center')

# Show the plot
plt.grid()
plt.show()
```

By observation, we can find that the crossover point occurs at n ≈ 15. This means that for input sizes less than 15 (e.g., n = 1, 2, ..., 15), Algorithm A is faster, and for input sizes greater than 15, Algorithm B is faster.

So, in summary, Algorithm B (3n log2(n)) is more efficient in terms of asymptotic time complexity, and the crossover point is approximately n ≈ 15.48, where Algorithm B becomes more efficient than Algorithm A as n increases.

## Problem 8 (10 Points)

Consider two algorithms, Algorithm A and Algorithm B, each with their respective time complexities as functions of the input size n:

Algorithm A requires $n^2$ seconds.
Algorithm B requires $2^n$ seconds.

Which algorithm grows asymptotically faster as n gets large, and what is the crossover value of n, if it exists?

## Solution:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions for Algorithm A and Algorithm B
def algorithm_A(n):
    return n**2

def algorithm_B(n):
    return 2**n

# Create an array of n values
n_values = np.arange(1, 20)

# Calculate the corresponding time complexities for both algorithms
time_complexity_A = [algorithm_A(n) for n in n_values]
time_complexity_B = [algorithm_B(n) for n in n_values]

# Plot the time complexities
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_complexity_A, label='Algorithm A (n^2)')
plt.plot(n_values, time_complexity_B, label='Algorithm B (2^n)')
plt.xlabel('Input Size (n)')
plt.ylabel('Time Complexity (T(n))')
plt.legend()
plt.grid(True)

# Find the crossover point (if it exists)
crossover_point = None
for n in n_values:
    if algorithm_A(n) < algorithm_B(n):
        crossover_point = n
        break

if crossover_point is not None:
    plt.axvline(x=crossover_point, color='red', linestyle='--', label=f'Crossover Point (n={crossover_point})')

plt.title('Comparison of Time Complexities')
plt.legend()
plt.show()

# Print the result
if crossover_point is not None:
    print(f"Algorithm B (2^n) grows asymptotically faster than Algorithm A (n^2), and the crossover point is at n = {crossover_point}.")
else:
    print("There is no crossover point; Algorithm B (2^n) always grows faster than Algorithm A (n^2).")
```
