## 1. Problem Statement - Maximal Network Rank

You are given an infrastructure of `n` cities connected by roads. Each road connects two cities bidirectionally, and this is represented as an array `roads` where `roads[i] = [ai, bi]`.

The **network rank** of two different cities is defined as the total number of directly connected roads to either city. If a road is directly connected to both cities, it is counted only once.

Your task is to find the **maximal network rank** of the entire infrastructure, which is the maximum network rank among all pairs of different cities.

Write a function `maximalNetworkRank(n, roads)` that takes an integer `n` (the number of cities) and a list of roads `roads`, and returns the maximal network rank.

### Constraints

- `2 <= n <= 100`
- `0 <= roads.length <= n * (n - 1) / 2`
- `roads[i].length == 2`
- `0 <= ai, bi <= n-1`
- `ai != bi`
- Each pair of cities has at most one road connecting them.


# Psuedo Code

```python
Function maximalNetworkRank(n, roads):
    Initialize connections as an array of size n, all initialized to 0.
    Initialize graph as a 2D array of size n x n, all initialized to False.
    
    For each road in roads:
        Extract cities 'a' and 'b' from the road.
        Increment connections[a] by 1  # Increase the connection count for city 'a'.
        Increment connections[b] by 1  # Increase the connection count for city 'b'.
        Set graph[a][b] = graph[b][a] = True  # Mark the road as connected in both directions.
    
    Initialize max_rank as 0  # Variable to store the maximum network rank.
    
    For i from 0 to n - 1:
        For j from i + 1 to n - 1:
            Initialize rank as connections[i] + connections[j]  # Calculate the initial rank.
            
            If graph[i][j] is True:  # Check if there's a direct road between city 'i' and 'j'.
                Decrement rank by 1  # Decrement the rank due to the shared road.
            
            Update max_rank as the maximum of max_rank and rank.
    
    Return max_rank  # Return the maximal network rank.

Call maximalNetworkRank(n, roads) to find the maximal network rank.

```

# Solution

````python
from typing import List

class Solution:
    def maximalNetworkRank(self, n: int, roads: List[List[int]]) -> int:
        # Initialize a list to keep track of the number of connections for each city.
        connections = [0] * n

        # Initialize a graph as a 2D array where each cell represents a road connection.
        graph = [[False] * n for _ in range(n)]

        # Iterate through the roads and update connections and the graph.
        for a, b in roads:
            connections[a] += 1  # Increase the connection count for city 'a'.
            connections[b] += 1  # Increase the connection count for city 'b'.
            graph[a][b] = graph[b][a] = True  # Mark the road as connected in both directions.

        # Initialize a variable to store the maximum network rank.
        max_rank = 0

        # Iterate through all pairs of cities.
        for i in range(n):
            for j in range(i + 1, n):
                rank = connections[i] + connections[j]  # Calculate the initial rank.

                # If there's a direct road between city 'i' and city 'j', decrement the rank.
                if graph[i][j]:
                    rank -= 1

                # Update the maximum network rank.
                max_rank = max(max_rank, rank)

        return max_rank
````

## Solution Breakdown

Okay, so let's talk about how we're gonna solve this problem. We want to find the biggest network rank in a bunch of connected cities with roads. Network rank is just a fancy way of saying how many roads are connected to a city.

**Step 1: Setting Up**
We're gonna start by getting organized. We'll create two things:
   - `connections`: A list to keep count of how many roads are connected to each city. We'll start with zeros for everyone.
   - `graph`: A 2D list to show which cities have direct roads. At first, everyone is not connected (marked as `False`).

**Step 2: Counting Connections**
Now, we'll go through all the roads we have and update our lists:
   - For each road `[a, b]`, we'll add 1 to the connection count for both city `a` and `b`.
   - We'll also mark `graph[a][b]` and `graph[b][a]` as `True` to show they have a direct road between them.

**Step 3: Find the Max Rank**
Here comes the fun part. We'll search for the highest network rank among all pairs of cities. We'll do this by:
   - Going through every pair of cities, like `i` and `j` where `i` is less than `j`.
   - We'll calculate an initial rank by adding up the connections for both cities, so `connections[i] + connections[j]`.
   - If there's a direct road between city `i` and `j`, we'll subtract 1 to avoid double-counting.
   - We'll keep track of the biggest rank we've seen so far in a variable called `max_rank`.

**Step 4: The Big Reveal**
Finally, we'll tell you the answer! We'll give you the `max_rank` we found, and that's the maximal network rank.

## Why It Works

   - We carefully set up our data to keep track of connections and roads.
   - We went through all possible pairs of cities and calculated the rank for each one.
   - `max_rank` always stores the highest rank we've seen.
   - So, when we return `max_rank`, it's gotta be the biggest network rank in the whole infrastructure.

# 2. Given a directed acyclic graph (DAG), implement a function to perform a topological sort.

```python
from collections import defaultdict

def topological_sort(graph):
    result = []
    visited = set()

    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor)
        result.append(node)

    for node in graph:
        if node not in visited:
            dfs(node)

    return result[::-1]

# Example
graph = {1: [2, 3], 2: [4], 3: [4], 4: []}
result = topological_sort(graph)
print(result)
```

**Example:**
Input: {1: [2, 3], 2: [4], 3: [4], 4: []}
Output: [1, 3, 2, 4]

**Proof of Correctness:**
The algorithm visits each node, explores its neighbors, and appends it to the result. The reversed result is a valid topological ordering. The correctness follows from the properties of DAGs.

**Complexity:**
Time: O(V + E) - where V is the number of vertices and E is the number of edges.
Space: O(V) - additional space for the visited set.


# 3.Implement a function to find the shortest path from a given source node to all other nodes in a weighted, directed graph using Dijkstra's algorithm.

```python
import heapq

def dijkstra(graph, source):
    distances = {node: float('infinity') for node in graph}
    distances[source] = 0
    priority_queue = [(0, source)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Example
graph = {1: {2: 1, 3: 4}, 2: {3: 2, 4: 5}, 3: {4: 1}, 4: {}}
result = dijkstra(graph, 1)
print(result)
```

**Example:**
Input: {1: {2: 1, 3: 4}, 2: {3: 2, 4: 5}, 3: {4: 1}, 4: {}}
Output: {1: 0, 2: 1, 3: 3, 4: 4}

**Proof of Correctness:**
Dijkstra's algorithm always selects the node with the smallest tentative distance, ensuring the correctness of the result.

**Complexity:**
Time: O((V + E) * log(V)) - where V is the number of vertices and E is the number of edges.
Space: O(V) - additional space for distances and the priority queue.

# 4. Given a connected, undirected graph, implement a function to find the minimum spanning tree using Kruskal's algorithm.

```python
def kruskal(graph):
    parent = {node: node for node in graph}

    def find_set(node):
        if parent[node] != node:
            parent[node] = find_set(parent[node])
        return parent[node]

    def union(u, v):
        root_u, root_v = find_set(u), find_set(v)
        parent[root_u] = root_v

    edges = [(weight, u, v) for u, neighbors in graph.items() for v, weight in neighbors.items()]
    edges.sort()

    mst = []
    for weight, u, v in edges:
        if find_set(u) != find_set(v):
            union(u, v)
            mst.append((u, v, weight))

    return mst

# Example
graph = {1: {2: 1, 3: 4}, 2: {3: 2, 4: 5}, 3: {4: 1}, 4: {}}
result = kruskal(graph)
print(result)
```

**Example:**
Input: {1: {2: 1, 3: 4}, 2: {3: 2, 4: 5}, 3: {4: 1}, 4: {}}
Output: [(1, 2, 1), (3, 4, 1), (2, 3, 2)]

**Proof of Correctness:**
Kruskal's algorithm selects the smallest edge at each step while avoiding cycles, ensuring that the result is a valid minimum spanning tree.

**Complexity:**
Time: O(E * log(V)) - where E is the number of edges and V is the number of vertices.
Space: O(V) - additional space for the parent data structure.


# 5. A manufacturing company has raw materials, manufacturing steps, and products, each with varying costs and sale prices. Implement a function to determine the process for the day that maximizes profit.

```python
import heapq

def max_profit_process(raw_materials, manufacturing_steps, products):
    graph = {}
    for material, cost in raw_materials.items():
        graph['source'] = graph.get('source', {})
        graph['source'][material] = cost

    for step, (material, product, cost) in enumerate(manufacturing_steps):
        graph[material] = graph.get(material, {})
        graph[material][product] = cost

    for product, sale_price in products.items():
        graph[product] = graph.get(product, {})
        graph[product]['sink'] = sale_price

    distances = {node: float('-infinity') for node in graph}
    distances['source'] = 0
    priority_queue = [(-float('infinity'), 'source')]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_distance < distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            profit = distances[current_node] - weight
            if profit > distances[neighbor]:
                distances[neighbor] = profit
                heapq.heappush(priority_queue, (-profit, neighbor))

    return distances['sink']

# Example
raw_materials = {'A': 2, 'B': 3}
manufacturing_steps = [('A', 'C', 1), ('B', 'C', 2), ('C', 'D', 3)]
products = {'D': 10}
result = max_profit_process(raw_materials, manufacturing_steps, products)
print(result)
```

**Example:**
Input:

Manufacturing Steps: [('A', 'C', 1), ('B', 'C', 2), ('C', 'D', 3)]
Products: {'D': 10}
Output: 8

**Proof of Correctness:**
The algorithm uses a single-source shortest path approach to find the maximum profit, considering costs and sale prices at each step.

**Complexity:**

Space: O(V) - additional space for distances and the priority queue.



# 6. Implement a function to find the shortest path from a given source node to a destination node in an unweighted, directed graph using Breadth-First Search (BFS).

```python
from collections import deque

def bfs_shortest_path(graph, source, destination):
    visited = set()
    queue = deque([(source, [])])

    while queue:
        current_node, path = queue.popleft()

        if current_node in visited:
            continue

        visited.add(current_node)
        path = path + [current_node]

        if current_node == destination:
            return path

        queue.extend((neighbor, path) for neighbor in graph[current_node] if neighbor not in visited)

    return None

# Example
graph = {1: [2, 3], 2: [4], 3: [4], 4: []}
result = bfs_shortest_path(graph, 1, 4)
print(result)
```

**Example:**
Input: {1: [2, 3], 2: [4], 3: [4], 4: []}
Output: [1, 3, 4]

**Proof of Correctness:**
BFS explores the graph level by level, ensuring that the first path found from the source to the destination is the shortest.

**Complexity:**
Time: O(V + E) - where V is the number of vertices and E is the number of edges.
Space: O(V) - additional space for the visited set and queue.

# 7. Implement a function to determine whether a given node is reachable from another node in a directed graph using Depth-First Search (DFS).

```python
def dfs_reachable(graph, start, target):
    visited = set()

    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor == target:
                return True
            if neighbor not in visited:
                if dfs(neighbor):
                    return True
        return False

    return dfs(start)

# Example
```

# 9. Given an undirected graph and two nodes A and B, implement a function to find the shortest path from node A to node B using Breadth-First Search.

```python
from collections import deque

def bfs_shortest_path(graph, start, end):
    queue = deque([(start, [start])])

    while queue:
        current, path = queue.popleft()

        if current == end:
            return path

        for neighbor in graph[current]:
            if neighbor not in path:
                queue.append((neighbor, path + [neighbor]))

    return None

# Example:
graph = {'A': ['B', 'C'], 'B': ['A', 'D', 'E'], 'C': ['A', 'F', 'G'], 'D': ['B'], 'E': ['B'], 'F': ['C'], 'G': ['C']}
result = bfs_shortest_path(graph, 'A', 'G')
print(result)
```

**Proof of Correctness:**
The algorithm explores nodes layer by layer, ensuring that the first path found is the shortest due to BFS properties.

**Complexity:**
Time: O(V + E) - where V is the number of vertices and E is the number of edges.



# 10. Implement a function to perform topological sorting on a directed acyclic graph (DAG) represented as an adjacency list.


```python
from collections import defaultdict, deque

def topological_sort(graph):
    in_degree = defaultdict(int)
    for node in graph:
        for neighbor in graph[node]:
            in_degree[neighbor] += 1

    queue = deque([node for node in graph if in_degree[node] == 0])
    result = []

    while queue:
        current = queue.popleft()
        result.append(current)

        for neighbor in graph[current]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    return result

# Example:
graph = {'A': ['B', 'C'], 'B': ['D'], 'C': ['D'], 'D': []}
result = topological_sort(graph)
print(result)
```

**Proof of Correctness:**
The algorithm ensures that each node is processed after its dependencies, adhering to the definition of topological sorting.

**Complexity:**
Time: O(V + E) - where V is the number of vertices and E is the number of edges.
