The problem can be solved using the "Stable Matching" algorithm, which ensures that no roommates would prefer each other over their current assignments.<br><br>
The algorithm looks like this:<br><br>
Initialize a list of unassigned roommates.<br>
While there are still unassigned roommates:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Find the roommate with the highest priority on the list of unassigned roommates.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Find the roommate with the highest priority on the preference list of the first roommate.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;If the second roommate is unassigned, assign them to each other.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Otherwise, if the second roommate prefers the first roommate to their current roommate, assign them &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;to each other and update the number of available positions in the first roommate's current room.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Otherwise, the first roommate remains in their current room.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;All roommates are now assigned<br><br>
Here's a step-by-step explanation of the algorithm:<br>

Step 1: Initialization<br>

Begin by having each roommate rank all other roommates according to their preferences.<br>
Create an array to keep track of each roommate's current assignment. Initially, no one is assigned, so all entries can be set to unassigned.<br><br>
Step 2: While There Are Unassigned Roommates<br>

Select an unassigned roommate, let's call this roommate "R1."<br>
R1 proposes to their most preferred roommate (the one they rank the highest and haven't proposed to yet). Let's call this proposed roommate "R2."<br>
If R2 is unassigned, or if R2 prefers R1 over their current roommate (if any), they accept the proposal, and both R1 and R2 become assigned to each other.<br>
If R2 is already assigned to someone and prefers their current roommate over R1, they reject the proposal, and R1 remains unassigned.<br>
Repeat these steps until there are no unassigned roommates left.<br><br>
Step 3: Result<br>

Once every roommate is assigned, you'll have a stable roommate assignment.<br><br>
Example:<br>

The Gale-Shapley theorem proves the correctness of this algorithm, which states that the algorithm always produces a stable matching. It guarantees that there are no pairs of roommates who would both prefer to live with each other over their current assignments.<br>

**Q5 (10 Points)
Given a set of items, each with a weight and a value, and a maximum weight capacity W, find the subset of items with the maximum total value such that the total weight of the subset does not exceed W.**

<u>Solution</u>:

Lets understand this problem with an example. Imagine having a backpack and we want to fill the backpack with some toys. Now each toy has different weight and different value. Our goal is to pick the toys that will have the most value without making our backpack too heavy to carry.<br>
Here's how we can solve this problem step by step:

Make a List: First, we make a list of all the toys we have, and for each toy, we write down how heavy it is and whats the value of toy.

Decide the Backpack Size: Next, we decide how big your backpack is, meaning it can only carry toys up to a certain total weight (let's call it W).

Pick the Best Toys: Now, we want to figure out which toys to put in your backpack. We start with an empty backpack and look at your list of toys. We pick one toy at a time and decide whether to put it in our backpack or not.

We look at the first toy on your list. If it's too heavy to fit in our backpack, we skip it and move to the next toy.<br>
If the toy can fit in your backpack, we compare it to the toys already in our backpack. If it's has a higher value than what's already inside, we put it in your backpack. If not, we skip it.

Repeat Until Full: Keep doing this for each toy on the list, one by one, until backpack is full. Make sure not to put too many toys so that backpack doesn't get too heavy.

You're Done: Once our backpack is full, we've picked the best combination of toys that give the most value without being too heavy to carry. 

Here is the code example for above algorithm:



```python
def knapsack(items, max_weight):
    # Create a table to store the best values for different weights
    dp = [[0 for _ in range(max_weight + 1)] for _ in range(len(items) + 1)]
    
    # Create a table to store the information about which items are included
    included = [[False for _ in range(max_weight + 1)] for _ in range(len(items) + 1)]

    # Fill in the table using a loop
    for i in range(1, len(items) + 1):
        for w in range(max_weight + 1):
            weight, value = items[i - 1]

            # If the current item's weight is more than the current weight limit,
            # we can't include it in the knapsack
            if weight > w:
                dp[i][w] = dp[i - 1][w]
            else:
                # We have two choices: include the current item or skip it
                # We choose the option that gives us the maximum value
                if dp[i - 1][w] >= dp[i - 1][w - weight] + value:
                    dp[i][w] = dp[i - 1][w]
                else:
                    dp[i][w] = dp[i - 1][w - weight] + value
                    included[i][w] = True

    # Find the items that are included in the optimal <u>Solution</u>
    selected_items = []
    i, w = len(items), max_weight
    while i > 0 and w > 0:
        if included[i][w]:
            selected_items.append(items[i - 1])
            w -= items[i - 1][0]
        i -= 1

    # Return the maximum value and the selected items
    return dp[len(items)][max_weight], selected_items

# Example usage:
toys = [(2, 5), (3, 8), (5, 13), (10, 22)]
max_weight = 10
max_value, selected_toys = knapsack(toys, max_weight)
print("Maximum value of toys in the backpack:", max_value)
print("Selected toys:", selected_toys)
```

**Q9 (25 Points) :: Coding Problem<br>
The annual National Science Fair is approaching, and there are students from various schools who wish to participate. Each student has a list of their preferred schools they want to attend for the fair, and each school has a list of preferred students. The fair organizers want to create a stable matching of students to schools using the Gale-Shapley algorithm.**

**A. Implement the Gale-Shapley algorithm in Python to match students to schools. You can create preference lists for each student and school.**

<u>Solution</u>:

```python
# Implement Gale-Shapley algorithm
def gale_shapley(student_prefs, school_prefs):
    # Initialize variables
    student_matches = {}
    school_assignments = {}

    while len(student_prefs) > 0:
        student = next(iter(student_prefs))
        school_choices = student_prefs[student]

        for school in school_choices:
            if school not in school_assignments:
                school_assignments[school] = student
                student_matches[student] = school
                del student_prefs[student]
                break
            else:
                current_match = school_assignments[school]
                if school_prefs[school].index(student) < school_prefs[school].index(current_match):
                    student_matches[current_match] = None
                    del student_prefs[current_match]
                    school_assignments[school] = student
                    student_matches[student] = school
                    del student_prefs[student]
                    break

    return student_matches

# Sample preference lists
student_preferences = {
    'Student1': ['School1', 'School2', 'School3'],
    'Student2': ['School2', 'School1', 'School3'],
    'Student3': ['School1', 'School3', 'School2']
}

school_preferences = {
    'School1': ['Student1', 'Student3', 'Student2'],
    'School2': ['Student2', 'Student1', 'Student3'],
    'School3': ['Student3', 'Student2', 'Student1']
}

matches = gale_shapley(student_preferences, school_preferences)
print(matches)
```

**B. To ensure fairness, shuffle the preference lists of both students and schools 1000 times and calculate the percentage of stable matches.**

<u>Solution</u>:

```python
import random

# Define student_preferences and school_preferences with varied preferences
student_preferences = {
    'Student1': ['School1', 'School2', 'School3'],
    'Student2': ['School2', 'School1', 'School3'],
    'Student3': ['School1', 'School3', 'School2']
}

school_preferences = {
    'School1': ['Student2', 'Student1', 'Student3'],
    'School2': ['Student1', 'Student2', 'Student3'],
    'School3': ['Student3', 'Student2', 'Student1']
}

# Implement Gale-Shapley algorithm
def gale_shapley(student_prefs, school_prefs):
    # Initialize variables
    student_matches = {}
    school_assignments = {}

    # Create copies of the keys
    student_keys = list(student_prefs.keys())
    
    while len(student_keys) > 0:
        student = student_keys.pop(0)  # Remove the first student from the list
        school_choices = student_prefs[student]

        for school in school_choices:
            if school not in school_assignments:
                school_assignments[school] = student
                student_matches[student] = school
                break
            else:
                current_match = school_assignments[school]
                if school_prefs[school].index(student) < school_prefs[school].index(current_match):
                    student_matches[current_match] = None
                    student_keys.append(current_match)  # Add the previous match back to the list
                    school_assignments[school] = student
                    student_matches[student] = school
                    break

    return student_matches

# Shuffle the preferences
def shuffle_preferences(preferences, num_times=1000):
    for i in range(num_times):
        for key in preferences:
            random.shuffle(preferences[key])

shuffle_preferences(student_preferences)
shuffle_preferences(school_preferences)

# Calculate percentage of stable matches (you can use the stable_matching function)
stable_matches = gale_shapley(student_preferences, school_preferences)
percentage_stable = len(stable_matches) / len(student_preferences) * 100
print(f"Percentage of stable matches: {percentage_stable}%")
```
