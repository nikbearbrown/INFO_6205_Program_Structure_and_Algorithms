    1.	Create a list of all the edges in the graph, along with their weights.

Input: A = [1, 3, 5, 7, 9], T = 6
Output: -1
Constraints:

A is a sorted array of integers.
T is an integer.

        Solution:
        The binary search algorithm works by recursively dividing the search range in two parts. At each step, the algorithm compares the target integer T to the middle element of the search range. If T is equal to the middle element, the algorithm returns the index of the middle element. Otherwise, the algorithm recursively searches the half of the search range that contains T (either left half or the right half)

        The algorithm terminates when the search range is empty, or when the algorithm finds the target integer T. In both case, the algorithm returns the correct index, or -1 if T is not found in the array.

````python
def binary_search(A, T, low, high):
  """Finds the index of the first occurrence of `T` in the sorted array `A` between indices `low` and `high`.

  Args:
    A: A sorted array of integers.
    T: A target integer.
    low: The low index of the search range.
    high: The high index of the search range.

  Returns:
    The index of the first occurrence of `T` in `A`, or `-1` if `T` is not found in `A`.
  """

  if low > high:
    return -1

  mid = (low + high) // 2

  if A[mid] == T:
    return mid

  elif A[mid] < T:
    return binary_search(A, T, mid + 1, high)

  else:
    return binary_search(A, T, low, mid - 1)


def find_first_occurrence(A, T):
  return binary_search(A, T, 0, len(A) - 1)
````

You are given a set of cities and the distances between them. Your task is to find the shortest path that visits each city exactly once and returns to the starting city. This is known as the Traveling Salesman Problem (TSP). However, there's a twist: you are required to find two disjoint paths that collectively cover all cities while minimizing the total distance.

Write a function minimize_distance(cities: List[Tuple[int, int]]) -> Tuple[List[Tuple[int, int]], List[Tuple[int, int]]] that takes a list of cities represented as (x, y) coordinates and returns two paths as lists of city indices. Each path should start and end at the same city, and no city should appear in both paths.

Your goal is to minimize the sum of the distances of the two paths.

Input:

cities: A list of tuples representing the coordinates of cities. The coordinates are integers where -1000 <= x, y <= 1000. The list contains at least two cities and at most 100 cities.
Output:

A tuple containing two lists of city indices, where each list represents a path.
Constraints:

The cities are unique; no city appears more than once.
You cannot visit the same city in both paths.
You should return the two paths such that the sum of their distances is minimized.

Example:

cities = [(0, 0), (1, 2), (2, 4), (3, 1), (4, 3)]
path1, path2 = minimize_distance(cities)

-> Output should be two paths with minimum total distance.\
 For example: 
            path1 = [0, 1, 3, 2, 4, 0]\
            path2 = [0, 2, 3, 1, 4, 0]

    Solution: This problem can be solved using dynamic programming and bit masking. We can create a memoization table where each state represents the current city and the set of visited cities so far. Then, we can iterate through all possible states and compute the minimum distance to reach that state.

        Below is a Python function that solves the problem:

````python
import itertools
import math

def calculate_distance(city1, city2):
    return math.sqrt((city1[0] - city2[0])**2 + (city1[1] - city2[1])**2)

def minimize_distance(cities):
    n = len(cities)
    all_cities = set(range(n))
    
    # Initialize the memoization table with large values
    memo = [[float('inf')] * n for _ in range(1 << n)]
    
    # Initialize the starting state
    memo[1][0] = 0
    
    # Calculate distances between all pairs of cities
    distances = [[calculate_distance(cities[i], cities[j]) for j in range(n)] for i in range(n)]
    
    # Iterate through all possible states
    for mask in range(1 << n):
        for u in range(n):
            if (mask >> u) & 1:
                for v in range(n):
                    if u != v and (mask >> v) & 1:
                        memo[mask][u] = min(memo[mask][u], memo[mask ^ (1 << u)][v] + distances[v][u])
    
    # Reconstruct the paths
    path1, path2 = [], []
    mask = (1 << n) - 1
    u = 0
    
    # Find the city that ends path 1
    v = min(range(1, n), key=lambda x: memo[mask][x] + distances[x][u])
    
    # Reconstruct path 1
    while mask:
        path1.append(u)
        mask ^= (1 << u)
        u, v = v, min(range(n), key=lambda x: memo[mask][x] + distances[x][u])
    
    # Reconstruct path 2
    path2 = [i for i in range(n) if i not in path1]
    
    return path1 + [0], path2 + [0]
````

You are given an array of integers, and your task is to find the maximum possible sum of a subarray. A subarray is a contiguous section of an array.

Write a function `max_subarray_sum(nums: List[int]) -> int` that takes in an array of integers `nums` and returns the maximum subarray sum.

Input:
- A list of integers `nums` (1 <= |nums| <= 10^5) where |nums| represents the length of the array. Each element of the array |nums[i]| (-10^4 <= |nums[i]| <= 10^4).

Expected Output:

- An integer, representing the maximum subarray sum.

**Example:**

```python
max_subarray_sum([1, 2, 3, 4, 5]) -> 15
max_subarray_sum([-2, 1, -3, 4, -1, 2, 1, -5, 4]) -> 6
max_subarray_sum([-1, -2, -3, -4, -5]) -> -1
```

Constraints:

Your solution should have a time complexity of O(n), where n is the length of the input array `nums`.

    Solution:

    We can solve this problem using Kadane's algorithm, which has a time complexity of O(n). Here's a pseudocode solution:

    Proof of Correctness:

    Kadane's algorithm works by iteratively updating `current_sum` to either the current element `num` or the sum of `current_sum` and `num`, whichever is greater. This ensures that `current_sum` always represents the maximum sum of a subarray ending at the current element.

    By maintaining `max_sum` as the maximum of all `current_sum` values encountered during the iteration, we guarantee that it stores the maximum subarray sum in the entire array. This algorithm finds the maximum subarray sum efficiently in O(n) time, as it only requires a single pass through the array.

````python
def max_subarray_sum(nums):
    max_sum = current_sum = nums[0]  # Initialize max_sum and current_sum to the first element of the array.

    for num in nums[1:]:  # Iterate through the array starting from the second element.
        current_sum = max(num, current_sum + num)  # Calculate the current sum.
        max_sum = max(max_sum, current_sum)  # Update max_sum if the current sum is greater.

    return max_sum
````

You are tasked with helping a robot navigate through a grid-based maze. The maze consists of cells arranged in rows and columns. Each cell can be either empty, represented by '0', or blocked, represented by '1'. The robot can move up, down, left, or right between adjacent cells. The goal is to find the shortest path from the starting cell to the target cell, if a valid path exists. If no valid path exists, the robot should return -1.

Write a Python function shortest_path(maze) that takes a 2D list maze as input, where maze[i][j] represents the content of cell (i, j). The function should return the length of the shortest path from the top-left corner (0, 0) to the bottom-right corner (n-1, m-1), where n is the number of rows, and m is the number of columns in the maze. If no valid path exists, return -1.

The robot can only move to adjacent cells if they are empty, and it cannot move diagonally. The robot can start from any cell in the maze, and the target is always the bottom-right cell.

Input:

A 2D list maze representing the grid (1 <= n, m <= 100).
Each element of the maze can be either '0' (empty cell) or '1' (blocked cell).
Output:

An integer representing the length of the shortest path from the top-left corner (0, 0) to the bottom-right corner (n-1, m-1).

Constraints:

The maze grid will have at most 100 rows and 100 columns.
The input maze is guaranteed to have a valid starting cell at (0, 0) and a valid target cell at (n-1, m-1).
The robot can only move up, down, left, or right; no diagonal movements are allowed.
The maze may contain obstacles ('1') that the robot cannot pass through.

    Solution:
    To solve this problem, we can use breadth-first search (BFS) to explore the maze. BFS is a suitable choice because it guarantees finding the shortest path when applied to a grid. Here's the pseudocode for the shortest_path function:


````python
from collections import deque

def shortest_path(maze):
    n = len(maze)
    m = len(maze[0])

    # Define the four possible movements: up, down, left, and right.
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    # Create a visited grid to keep track of visited cells.
    visited = [[False for _ in range(m)] for _ in range(n)]

    # Initialize a queue for BFS.
    queue = deque([(0, 0, 0)])  # (row, col, distance)

    while queue:
        row, col, distance = queue.popleft()

        # Check if we've reached the target cell.
        if row == n - 1 and col == m - 1:
            return distance

        for dr, dc in moves:
            r, c = row + dr, col + dc

            # Check if the new cell is within the maze boundaries and is not visited.
            if 0 <= r < n and 0 <= c < m and maze[r][c] == '0' and not visited[r][c]:
                visited[r][c] = True
                queue.append((r, c, distance + 1))

    # If we reach here, there's no valid path.
    return -1
````

    The BFS algorithm explores the maze while keeping track of the distance from the starting cell. It considers only empty cells ('0') and ensures that it finds the shortest path.

Further, lets suppose that despite the bucolic setting, the residents of all these houses are avid cell phone users. You want to place cell phone base stations at certain points along the road, so that every house is within six miles of one of the base stations.
Give an efficient algorithm along with pseudo code that achieves this goal, using as few base stations as possible

    Solution:
    This problem can be efficiently solved using a greedy algorithm. The idea is to place base stations at strategic locations that minimize the number of base stations while ensuring that every house is within six miles of at least one base station. Here's the algorithm along with pseudocode:

    Algorithm: Place Cell Phone Base Stations

    1. Initialize an empty list base_stations to store the positions of the base stations.

    2. Sort the list of house positions in ascending order.

    3. Initialize a variable current_position to the western endpoint of the road.

    4. Iterate through the sorted list of house positions:

        i. For each house, check if it is more than six miles away from current_position. If it is,
           place a base station at the position of the previous house (i.e., current_position) + six miles.
        ii. Update current_position to the position of the newly placed base station.
            
    5. Return the list base_stations containing the positions of the base stations.

````python
def place_base_stations(house_positions):
    # Sort house positions in ascending order
    sorted_houses = sorted(house_positions)
    
    # Initialize variables
    base_stations = []
    current_position = sorted_houses[0]  # Initialize with the first house
    
    # Place the first base station
    base_stations.append(current_position + 6)
    
    for house in sorted_houses:
        # Check if the current house is more than six miles away
        if house > current_position + 6:
            # Place a base station and update the current position
            base_stations.append(house + 6)
            current_position = house + 6
    
    return base_stations

# Example usage:
house_positions = [1, 9, 20, 25, 30, 40]
base_stations = place_base_stations(house_positions)
print("Optimal base station positions:", base_stations)
````

    The below code visualizes Dijkstra's algorithm on a simple graph. W can customize and expand it for our specific use case or graph structure. We will have to adjust visualization parameters, such as colors, labels, and positions, to suit our needs.

````python
import matplotlib.pyplot as plt
import networkx as nx
import heapq

# Define your graph here as an adjacency list
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}

# Dijkstra's Algorithm
def dijkstra(graph, start, end):
    shortest_path = {}
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_node in shortest_path:
            continue
        
        shortest_path[current_node] = current_distance
        
        for neighbor, weight in graph[current_node].items():
            if neighbor not in shortest_path:
                heapq.heappush(priority_queue, (current_distance + weight, neighbor))
    
    return shortest_path

# Visualization code
G = nx.Graph()

# Add nodes and edges to the graph
for node in graph:
    G.add_node(node)
    for neighbor, weight in graph[node].items():
        G.add_edge(node, neighbor, weight=weight)

# Compute shortest path using Dijkstra's algorithm
shortest_path = dijkstra(graph, 'A', 'D')

# Visualization code
pos = nx.spring_layout(G)  # Position nodes using a spring layout
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
node_colors = ['green' if node == 'A' else 'red' if node == 'D' else 'gray' for node in G.nodes()]

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=500, font_size=12, font_color='white')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue')
plt.title('Dijkstra\'s Algorithm Visualization')
plt.show()

print("Shortest Path:", shortest_path['D'])
````

````python
import matplotlib.pyplot as plt
from collections import defaultdict

# Define your graph here (as an adjacency list)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Depth-First Search to check if A is reachable from B
def dfs(graph, start, target, visited):
    visited.add(start)
    # Visualization code: Plot the visited node
    plt.scatter(node_positions[start][0], node_positions[start][1], color='red', s=100)
    plt.pause(0.5)  # Add a small delay to visualize the process
    
    if start == target:
        return True
    
    for neighbor in graph[start]:
        if neighbor not in visited:
            if dfs(graph, neighbor, target, visited):
                return True
    
    return False

# Visualization code: Create node positions
node_positions = {'A': (0, 0), 'B': (1, 1), 'C': (1, -1), 'D': (2, 2), 'E': (2, 0), 'F': (2, -2)}

# Visualization code: Plot the graph
for node, pos in node_positions.items():
    plt.text(pos[0], pos[1], node, fontsize=12, ha='center', va='center')
for node, neighbors in graph.items():
    for neighbor in neighbors:
        plt.plot([node_positions[node][0], node_positions[neighbor][0]],
                 [node_positions[node][1], node_positions[neighbor][1]], 'k-')

plt.axis('equal')
plt.gca().invert_yaxis()
plt.title('DFS Visualization')
plt.show()

# Call DFS to check if A is reachable from B
start_node = 'A'
target_node = 'B'
visited_nodes = set()
reachable = dfs(graph, start_node, target_node, visited_nodes)

if reachable:
    print(f"{start_node} is reachable from {target_node}")
else:
    print(f"{start_node} is not reachable from {target_node}")
````
