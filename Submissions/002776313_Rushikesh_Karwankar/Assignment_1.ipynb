**Solution 3:** 
Solution using the Gale-Shapley Algorithm for Hospital Resident Matching:

The Gale-Shapley algorithm, also known as the Deferred Acceptance algorithm, provides a systematic way to solve the hospital resident matching problem. Here is how the algorithm works:

Initialization: Start with all medical students and hospitals being unmatched.

Iteration: Repeat the following steps until no further changes occur:

Each unmatched student proposes to their most preferred hospital. Students make their proposals based on their own preference order.
Each hospital receives proposals and tentatively accepts the most preferred students, up to their capacity. If a hospital is already filled to capacity, it accepts the proposing student only if they are more preferred than its least preferred current resident.
Rejected students remain unmatched and will propose to their next most preferred hospital in the next iteration.
Termination: The algorithm terminates when no more proposals can be made or accepted.

Result: The final matching is considered stable because no student-hospital pair would both prefer each other to their assigned match.

Proof of Stability:
The Gale-Shapley algorithm guarantees a stable matching. To prove this, consider the following:

Suppose there is an unstable pair (S, H) where S is a student and H is a hospital who both prefer each other to their current assignments. This means that either S proposed to H before and was rejected or H tentatively accepted S at some point but later replaced S with another student they preferred more.

In the first case, if S proposed to H before, it implies that H either never preferred S more than their current resident or they changed their preference during the algorithm. But since the algorithm only accepts a new student if they are strictly more preferred, this is not possible.

In the second case, if H tentatively accepted S but later replaced them, it means H found a student they preferred more than S. However, this also cannot occur since the algorithm always selects the most preferred students for hospitals.

Therefore, there cannot be any unstable pairs in the final matching, and it is a stable solution.

The Gale-Shapley algorithm also ensures that the resulting matching is efficient and respects the preferences of both students and hospitals. It optimizes for the preferences of students while ensuring that hospitals receive residents according to their own preference rankings, making it a fair and effective solution to the hospital resident matching problem.

**Pseudocode Representation:**

```plaintext
function hospital_resident_matching(students, hospitals):
    # Initialize all students and hospitals as unmatched
    student_matching = {}
    hospital_matching = {}

    while there are unmatched students:
        for each student in students:
            if student is not already matched:
                hospital = student's most preferred hospital
                if hospital is not full:
                    student_matching[student] = hospital
                    hospital_matching[hospital] = student
                else:
                    current_resident = the least preferred resident at hospital
                    if student is more preferred than current_resident:
                        student_matching[student] = hospital
                        hospital_matching[hospital] = student
                        student_matching[current_resident] = None

    return student_matching
```

from collections import deque

def is_connected(graph, user1, user2):
    def bfs(start, end):
        visited = set()
        queue = deque()
        parent = {}  # To track the parent of each node in the path
        
        queue.append(start)
        visited.add(start)
        
        while queue:
            current_user = queue.popleft()
            if current_user == end:
                path = []
                while current_user is not None:
                    path.append(current_user)
                    current_user = parent.get(current_user)
                return list(reversed(path))
            
            for neighbor in graph.get(current_user, []):
                if neighbor not in visited:
                    queue.append(neighbor)
                    visited.add(neighbor)
                    parent[neighbor] = current_user
        
        return False  # No path found
    
    if user1 == user2:
        return [user1]
    
    path = bfs(user1, user2)
    
    if path:
        return path
    else:
        return False

def dfs_delivery_route(graph, node, visited, package_capacity, time_window, penalty):
    if not visited[node]:
        visited[node] = True
        best_route = None
        best_time = float('inf')

        for neighbor, delivery_time in graph[node]:
            if not visited[neighbor]:
                current_route, current_time = dfs_delivery_route(graph, neighbor, visited, package_capacity, time_window, penalty)
                if current_time < best_time and current_route.count(1) <= package_capacity:
                    best_route = [1] + current_route
                    best_time = current_time

        if best_route is None:
            return [0], 0  # No more deliveries to make from this location
        else:
            return best_route, best_time + penalty if best_time > time_window else best_time

def optimize_delivery_route(locations, graph, package_capacity, penalty):
    visited = [False] * len(locations)
    best_route = None
    best_time = float('inf')

    for i, location in enumerate(locations):
