```python
#Creating Custom Object Int 

class Int():
    # Constructor (Data Strcuture)
    def __init__(self,n:'int'  = 0):  
        self._positive=True
        if(n<0):
            self._positive=False
        self._a = self.build(n)
    #Overloading prebuild python functions
        #overloading '__str__' to get print function to work on custom INT (Data Strcuture)
    def __str__(self):
        n=self.int()
        sign="-"
        if(self._positive):
            sign="+"
        s = str(n) + " " + sign + "["
        d = len(self)
        for i in range(d):
            x = self[i]
            s=s+str(x)
            if(i< d-1) :
                s=s+", "
        s=s+"]"
        return s
        #overloading 'len' to get length of custom INT (Data Strcuture)
    def __len__ (self)->'int':
        return len(self._a)
        ##overloading '__getitem' for a[]= to work (Data Strcuture)
    def __getitem__(self,pos)->'int':
        n = len(self)
        assert(pos>= 0 and pos <n)
        return self._a[pos] 
         ##overloading '__getitem' for a[]= to work (Data Strcuture)
    def __setitem__(self,pos,v):
        n = len(self)
        assert(pos >= 0 and  pos < n )
        self._a[pos] = v 
         ##overloading '__add__' for addition of Int (Aritmatic)
    def __add__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        vc=va+vb
        c=Int(vc)
        return c
        ##overloading 'sub' for Substraction of Int (Arithmatic)
    def __sub__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        vc=va-vb
        c=Int(vc)
        return c
        #overloading 'greaterThan' for relational  operation of Int (Relational)
    def __gt__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        return(va>vb)
        #overloading 'lessThan' for relational  operation of Int (Relational)
    def __lt__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        return(va<vb)
        #overloading 'equal' for relational  operation of Int (Relational)
    def __eq__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        return(va==vb)
        #overloading 'notequal' for relational  operation of Int (Relational)
    def __ne__(self,b:'Int'):
        va=self.int()
        vb=b.int()
        return(va!=vb)
        
 #Programmer Defined functions
    #Given a an integer convert it to a list of integer (Data Strcuture)
    def build(self,n:'int')->'list': 
        if(n<0):
            n=-n
        l=[]
        if(n<10):
            l.append(n)
        else:
            while (n!=0):
                l.append(n%10)
                n=n//10
        self._reverse(l)
        return l

    #reverse the integers in the given list (Data Strcuture)
    def _reverse(self,l):
        i=0
        j=len(l)-1
        while(i<j):
            t=l[i]
            l[i]=l[j]
            l[j]=t
            i=i+1
            j=j-1
    #convert list into integer (Access)
    def int(self):
        v = 0
        for e in self._a:
            v = 10 * v + e
        if(v== 0 or self._positive):
            return v
        return -v
    
```
